## indexing & slicing

* 1차원 배열의 indexing & slicing : **`[]`**
* 2차원 배열의 indexing & slicing : **`[,]`**
* 3차원 배열의 indexing & slicing : **`[,,]`**

```python
arr = np.array([1,2,3,4,5])
print('arr>>>', arr, '\n')

# ----------------------------
# 인덱싱
# ----------------------------
print(arr[0]) #1
print(arr[1]) #2
print(arr[2]) #3
print(arr[3]) #4
print(arr[4]) #5
```

## fancy indexing
* 인덱스 배열을 이용하여 인덱싱

```python
arr = np.array([100,101,102,103,104])
print('arr>>>\n', arr)
print('--------------------')
idx = [0, 2, 4]
print('0,2,4번 인덱싱 :', arr[idx])
```

## boolean indexing
* boolean 배열을 이용하여 인덱싱
* 조건 필터링
    * `arr[cond]`
    * `np.where(cond, x, y)` 
        * x:True일 때 값
        * y:False일때 값

```python
arr = np.array([100,101,102,103,104])
bool_idx = [True, False, True, False, True]

print('arr>>>\n', arr, '\n')
print('bool_index>>>\n', bool_idx, '\n')

print('----- boolean indexing -----')
print(arr[bool_idx])
```

```python
nums = np.array([100,103,102,101,109])
cond = (nums % 2 != 0)
nums[cond]
```

```python
# -----------------
# 홀수만 선택하기
# -----------------

arr = np.array([100,103,102,101,109])
print('arr>>>\n', arr)
print('---------------')

# ------------------------------------------
# 홀수만 선택하기 위한 bool_index 생성
# ------------------------------------------
cond = (arr%2 == 1)

print('bool_idx>>>\n', cond)

print('----- boolean indexing -----')
print(arr[cond])
```

```python
# -----------------
# 홀수, 짝수 여부 배열 만들기 
# ['짝' '홀' '짝' '홀' '홀']
# -----------------

arr = np.array([100,103,102,101,109])
print('arr>>>\n', arr)
print('---------------')

cond = arr % 2 == 1
print(cond)


print(np.where(cond, '홀', '짝'))
```

## 배열 복사
* `배열객체.copy()`
* 슬라이싱 된 배열은 원본 배열을 참조
    * 슬라이싱 된 배열의 내용 수정 시 원본 배열이 수정됨
    * copy하여 사용할 것을 권장.

```python
# -----------------
# 슬라이싱 된 배열 수정
# -----------------

arr1 = np.array([1,2,3,4,5])
arr2 = arr1[:3]

print('arr1:',arr1)
print('arr2(arr1에서 슬라이싱):',arr2, '\n')

print('--------arr2[0]을 100으로 수정하고 arr1, arr2 출력---------')
arr2[0] = 100
print('arr1:', arr1)
print('arr2:', arr2)
```

```python
# -----------------
# 슬라이싱 된 배열을 copy본으로 만들고 수정
# -----------------

arr1 = np.array([1,2,3,4,5])
arr2 = arr1[:3].copy()

print('arr1:',arr1)
print('arr2(arr1에서 슬라이싱한 copy본):',arr2, '\n')

print('--------arr2[0]을 100으로 수정하고 arr1, arr2 출력---------')
arr2[0] = 100
print('arr1:', arr1)
print('arr2:', arr2)
```
