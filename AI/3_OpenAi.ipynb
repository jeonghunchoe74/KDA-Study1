# 1. 텍스트를 입력 받는다.
# 2. 입력받은 텍스트가 프롬프트가 된다.
# 3. LLM에 입력받은 텍스트로 질문을 한다.
# 4. 받은 내용을 화면에 보여준다.


```python
messages_history=[]
while True :
    text = input("질문을 입력하세요 : ")
    if text.lower() == 'ㅂ' or text.lower() == 'q':
        break

    messages_history.append({"role":"user", "content":text})  # 1번째 user 질문
    ret = client.chat.completions.create(
        model= default_model,
        messages=messages_history
    )

    print(ret.choices[0].message.content)

    messages_history.append({"role":"assistant", "content":ret.choices[0].message.content}) #LLM의 답변
```

# gradio로 챗봇 만들기

```python
import gradio as gr

# 인터페이스에서 이벤트가 발생되면 실행되는 함수
def func(text):
    return f"Hello {text}!"

demo = gr.Interface(fn=func, inputs="text", outputs="text")

demo.launch() #server_name="0.0.0.0"
```

```python
import random
import time
def chat_response(msg, chat_history):
    messages_history=[]
    for user_msg, assistant_msg in chat_history:
        messages_history.append({"role": "user", "content": user_msg})
        if assistant_msg:
            messages_history.append({"role": "assistant", "content": assistant_msg})
    messages_history.append({"role":"user", "content":msg})
    print(messages_history)
    ret = client.chat.completions.create(
        model=default_model,
        messages = messages_history,
        stream = True
    )
    total_stream = ""
    for char in ret:
        ret_stream = getattr(char.choices[0].delta, "content", None)
        if ret_stream:
            total_stream += ret_stream
            yield chat_history + [(msg, total_stream)], ""
with gr.Blocks() as demo:
    chatbot = gr.Chatbot()
    msg = gr.Textbox(placeholder="질문을 입력하세요", label="질문")
    btn = gr.Button("초기화")
    msg.submit(fn=chat_response,
            inputs=[msg, chatbot],
            outputs=[chatbot, msg])
    btn.click(lambda:[], inputs=None, outputs=chatbot)
demo.launch()
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
video_id = "uzTLm7PRy30"
ytt = YouTubeTranscriptApi()
transcript = ytt.fetch(video_id, languages=['ko', 'en'])
for snippet in transcript:          # snippet은 FetchedTranscriptSnippet 객체
    print(snippet.text)             # ['text']가 아니라 .text
full_text = " ".join(s.text for s in transcript)
print("\n\nFull Transcript:\n", full_text)
```
