# Langchain

```python
from dotenv import load_dotenv
import os

load_dotenv(override=True)
api_key = os.getenv("OPENAI_API_KEY")
default_model = os.getenv("OPENAI_DEFAULT_MODEL")
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model=default_model, api_key=api_key)
llm
```

# invoke 단일 요청을 처리하는 방법, 가장 단순한 질문, 간단하고 직관적인 사용을 위해 설계됨

```python
messages = [("system", "당신은 영화 평론가입니다"), ("human", "최근 영화중 볼만한걸 추천해줘")]
response = llm.invoke(messages)
print(response.content)
```

# stream
# 실시간으로 데이터를 전송하는 방법
# 모델이 응답을 만들 동안 그 응답을 조각으로 나누어서 클라이언트 실시간으로 전달하는 방법

```python
messages = [("system", "당신은 서울의 음식과 문화 전문가입니다"), ("human", "서울에서 데이트할 때 가볼만한 분위기 좋은 곳들을 알려주세요")]

response = llm.stream(messages)
print(response)

for chunck in response:
    print(chunck.content, end="")
```

# batch
# 여러 요청을 한 번에 처리하는 것

```python
messages = [[("system", "당신은 서울의 음식과 문화 전문가입니다"), ("human", "서울을 대표하는 음식을 맛볼 수 있는 음식점 5개를 추천해주세요")],
            [("system", "당신은 서울의 음식과 문화 전문가입니다"), ("human", "서울에서 연인끼리 갈 수 있는 바를 알려주세요")]]

responses = llm.batch(messages)
for response in responses:
    print(response.content)
```

# Chat

```python
from langchain.schema import SystemMessage, HumanMessage
sys = SystemMessage(content="당신은 음악을 추천해 주는 전문 AI입니다")
user = HumanMessage(content="1990년대 재즈곡 5개만 추천해줘")
user1 = HumanMessage(content="1980s년대 록음악 5개만 추천해줘")
myMsg = llm([sys, user])
print(myMsg.content)
myMsg = llm([sys, user1])
print(myMsg.content)
```

```python
import asyncio
import time

messages = [("system", "당신은 서울의 음식과 문화 전문가입니다"), ("human", "서울을 대표하는 음식을 맛볼 수 있는 음식점 5개를 추천해주세요")]
           

async def invoke_async(llm, messages):
    response = await llm.ainvoke(messages)
    print(response.content)
async def invoke_parallel(llm, messages):
    tasks = [invoke_async(llm, messages) for _ in range(10)]
    await asyncio.gather(*tasks)
```

```python
print("Async")
start = time.perf_counter()
await invoke_parallel(llm, messages)
end = time.perf_counter()
print(f"Elapsed time: {end - start:.2f} seconds")
```

```python
print("Sync")
start = time.perf_counter()
for i in range(10):
    response = llm.invoke(messages)
    print(response.content)
    print()
end = time.perf_counter()
print(f"Elapsed time: {end - start:.2f} seconds")
```

# 메세지 append

```python
messages = [("system", "당신은 서울의 음식과 문화 전문가입니다"), ("human", "서울을 대표하는 음식을 맛볼 수 있는 음식점 5개를 추천해주세요")]

print('before----->', messages)
print("")
response = llm.invoke(messages)
messages.append(response)
print('after------->', messages)
```

# ChatMessageHistory

```python
from langchain.memory import ChatMessageHistory
memory = ChatMessageHistory()

#system 메세지
memory.add_message(SystemMessage(content="당신은 여행 전문가로 고객의 여행 일정에 도움을 줍니다"))

#user 메세지
memory.add_user_message("부산 여행에서 딱 한 곳만 가봐야 한다면 어떤 곳을 가야할지 추천해 주세요")

response = llm.invoke(memory.messages)
memory.add_ai_message(response.content)

for msg in memory.messages:
    print(msg)
```

```python
#user 메세지
memory.add_user_message("부산역에서 그 곳에 가는 교통편을 알려주세요")

response = llm.invoke(memory.messages)
memory.add_ai_message(response.content)

for msg in memory.messages:
    print(msg)
```

- ConversationBufferMemory -

```python
from langchain.memory import ConversationSummaryBufferMemory

memory = ConversationSummaryBufferMemory(llm=llm, return_messages=False, max_token_limit = 10) # 충분한 크기로 설정
memory.clear()

memory.save_context({'human':'안녕하세요'}, 
                    {'ai': '안녕하세요! 호칭을 어떻게 하는게 좋을까요?'})

memory.save_context({'human':'나는 홍길동입니다'}, 
                    {'ai': '네, 홍길동님 만나서 반갑습니다'})

memory.save_context({'human':'경복궁까지 가려면 어떻게 가는지 알려주세요'}, 
                    {'ai': '물론이죠. 경복궁까지 가려면 지하철을 타고 가시면 됩니다.'})
result = memory.load_memory_variables({})
print(result['history'])
```

```python
from langchain.memory import ConversationBufferWindowMemory

memory = ConversationBufferWindowMemory(memory_key='chat_history', return_messages=True, k=2)
memory.clear()

memory.save_context({'human':'안녕하세요'}, 
                    {'ai': '안녕하세요! 호칭을 어떻게 하는게 좋을까요?'})

memory.save_context({'human':'나는 홍길동입니다'}, 
                    {'ai': '네, 홍길동님 만나서 반갑습니다'})

memory.save_context({'human':'경복궁까지 가려면 어떻게 가는지 알려주세요'}, 
                    {'ai': '물론이죠. 경복궁까지 가려면 지하철을 타고 가시면 됩니다.'})
result = memory.load_memory_variables({})
for msg in result['chat_history']:
    print(msg)
```

```python
from langchain.callbacks import get_openai_callback
with get_openai_callback() as cb:
    messages = [HumanMessage(content="서울역에서 명동까지 가려면 어떻게 가야 하나요?")]
    response = llm.invoke(messages)
    print(response.content)

    print(cb.total_tokens)
    print(cb.prompt_tokens)
    print(cb.completion_tokens)
    print(cb.total_cost)
```
