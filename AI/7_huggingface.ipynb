```python
from langchain_openai import ChatOpenAI  
from langchain.agents import AgentExecutor, create_react_agent  
from langchain_core.prompts import PromptTemplate
from langchain_core.tools import Tool  
import openai 
import os


chat = ChatOpenAI(
    temperature=0.7, 
    max_tokens=2048, 
    model="gpt-4-turbo"  
)

# DALL·E 함수를 직접 정의
def generate_image(prompt: str) -> str:
    """DALL·E를 사용하여 이미지 생성"""
    try:
        client = openai.OpenAI()
        response = client.images.generate(
            model="dall-e-3",
            prompt=prompt,
            size="1024x1024",
            quality="standard",
            n=1,
        )
        image_url = response.data[0].url
        return f"이미지가 생성되었습니다. URL: {image_url}"
    except Exception as e:
        return f"이미지 생성 중 오류 발생: {str(e)}"

# Tool 객체로 래핑
dalle_tool = Tool(
    name="DALL-E Image Generator",
    description="텍스트 설명을 바탕으로 이미지를 생성합니다. 사용법: 생성하고 싶은 이미지의 상세한 설명을 입력하세요.",
    func=generate_image
)
tools = [dalle_tool]

# 프롬프트 템플릿 정의
prompt = PromptTemplate.from_template("""
당신은 DALL·E를 사용하여 이미지를 생성하는 전문가입니다.
사용자의 요청에 따라 적절한 이미지를 생성해주세요.

사용 가능한 도구:
{tools}

도구 이름: {tool_names}

다음 형식을 사용하세요:

Question: 답해야 할 질문
Thought: 무엇을 해야 하는지 생각해보세요
Action: 취할 행동 [{tool_names} 중 하나]
Action Input: 행동의 입력
Observation: 행동의 결과
... (이 Thought/Action/Action Input/Observation을 필요한 만큼 반복)
Thought: 이제 최종 답을 알았습니다
Final Answer: 원래 질문에 대한 최종 답

시작하세요!

Question: {input}
Thought: {agent_scratchpad}
""")


agent = create_react_agent(chat, tools, prompt)

agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    verbose=True,
    max_iterations=3,
    handle_parsing_errors=True
)

# 이미지 생성 명령 실행
try:
    result = agent_executor.invoke({
        "input": response
    })
    print("\n" + "="*50)
    print(result["output"])
    
except Exception as e:
    print(f"오류 발생: {str(e)}")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

pretrained_model = 'textattack/bert-base-uncased-imdb' #긍정/부정 2개의 레이블
# 1. 토크나이저와 분류 모델 불러오기 (3개의 레이블 분류 예시)
tokenizer = AutoTokenizer.from_pretrained(pretrained_model)
model = AutoModelForSequenceClassification.from_pretrained(pretrained_model, 
                                                        num_labels=2)

# 2. 입력 문장들 (배치 처리)
sentences = ["I love this product!", 
            "This is the worst movie ever.", 
            "It was okay, nothing special."]

# 3. 토큰화 (패딩과 트렁케이션 적용)
inputs = tokenizer(sentences, 
                padding=True, 
                truncation=True, 
                return_tensors="pt")

# 4. 모델에 입력하여 분류 결과(로짓) 얻기
outputs = model(**inputs)
logits = outputs.logits  # (batch_size, num_labels)

# 5. 확률로 변환 (선택사항)
probs = torch.softmax(logits, dim=-1)

print("로짓:", logits)
print("확률:", probs)

predicted_class_indices = torch.argmax(logits, dim=-1)  # [0, 1, 1] 처럼 나옴 
print(predicted_class_indices)
#인덱스 0은 부정, 1은 긍정으로 매핑하는 규칙을 적용
```

- Fine Tuning

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

pretrained_model = 'bert-base-cased' #긍정/부정 2개의 레이블
# 1. 토크나이저와 분류 모델 불러오기 (3개의 레이블 분류 예시)
tokenizer = AutoTokenizer.from_pretrained(pretrained_model)
model = AutoModelForSequenceClassification.from_pretrained(pretrained_model, 
                                                        num_labels=3)
```

```python
# 정답
dic = {0:'긍정', 1:'중립', 2:'부정'}

# 입력 문장
sentences = ["I like apple",
             "I like pear",
             "I go to scholl",
             "I hate banana",
             "I felt very tired",
             "I feel so good"]

answers = torch.tensor([0,0,1,2,2,0]) # 정답 레이블
```

```python
model.eval()

with torch.no_grad():
    for sentence in sentences:
        input = tokenizer.encode(sentence, return_tensors="pt")
        output = model(input)
        no = output.logits.argmax(dim=-1)
        print(dic[no.item()] , ':', sentence)
    
```

## 파인튜닝(에포크 : 50, 최적화함수: AdamW, 학습률:0.00001)

```python
loss = torch.nn.CrossEntropyLoss() # 다중 분류할 때 사용하는 손실함수
optimizer = torch.optim.AdamW(model.parameters(), lr=0.00001) # 학습률
```

```python
losses = []
model.train()
for epoch in range(50):
    # 그래디언트(기울기) 초기화
    optimizer.zero_grad()

    inputs = tokenizer.batch_encode_plus(sentences,
                                padding=True,
                                truncation=True,
                                max_length=512,
                                return_tensors="pt")
    outputs = model(**inputs)
    loss_rate = loss(outputs.logits, answers)
    loss_rate.backward()
    optimizer.step()
    losses.append(loss_rate.item())
    print('---------', epoch, loss_rate.item())
```

```python
import matplotlib.pyplot as plt

# 1. 그래프 전체 크기 설정
plt.figure(figsize=(10, 5))

# 2. x축은 에포크(epoch), y축은 손실(loss) 값으로 그래프 그리기
# marker='o'는 각 데이터 포인트를 점으로 표시해줍니다.
plt.plot(losses, marker='o', linestyle='-')

# 3. 그래프 제목 및 축 레이블 설정 (한글 깨짐 방지를 위해 영어로 작성)
plt.title('Training Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')

# 4. 그리드(격자) 표시
plt.grid(True)

# 5. 그래프 보여주기
plt.show()
```

```python
model.eval()

with torch.no_grad():
    for sentence in sentences:
        input = tokenizer.encode(sentence, return_tensors="pt")
        output = model(input)
        no = output.logits.argmax(dim=-1)
        print(dic[no.item()] , ':', sentence)
    
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# 사전학습된 언어 모델 불러옴 - 영어 기반 BERT 모델
pretrained_model = 'bert-base-cased'

# 모델 및 토크나이저 불러오기 
tokenizer = AutoTokenizer.from_pretrained(pretrained_model)
model = AutoModelForSequenceClassification.from_pretrained(pretrained_model, 
                                        num_labels=5) #3개의 레이블로 분류
```

```python
# 정답
dic = {0:'매우 부정', 1:'부정', 2:'중립', 3:'긍정', 4:'매우 긍정'}

# 입력 문장
sentences = ["I like apple",
             "I really hate pear",
             "I go to the movie",
             "I hate banana",
             "I felt very happy",
             "I feel good"]

answers = torch.tensor([3,0,2,1,4,3]) # 정답 레이블
```

```python
model.eval()

with torch.no_grad():
    for sentence in sentences:
        input = tokenizer.encode(sentence, return_tensors="pt")
        output = model(input)
        no = output.logits.argmax(dim=-1)
        print(dic[no.item()] , ':', sentence)
    
```

```python
loss = torch.nn.CrossEntropyLoss() # 다중 분류할 때 사용하는 손실함수
optimizer = torch.optim.AdamW(model.parameters(), lr=0.00001) # 학습률
```

```python
losses = []
model.train()
for epoch in range(50):
    # 그래디언트(기울기) 초기화
    optimizer.zero_grad()

    inputs = tokenizer.batch_encode_plus(sentences,
                                padding=True,
                                truncation=True,
                                max_length=512,
                                return_tensors="pt")
    outputs = model(**inputs)
    loss_rate = loss(outputs.logits, answers)
    loss_rate.backward()
    optimizer.step()
    losses.append(loss_rate.item())
```
