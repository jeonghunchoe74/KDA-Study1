```python
def low_per_pbr(year):
    df1 = stock.get_market_cap_by_ticker(f"{year}0101")
    df1 = df1[["종가", "시가총액"]]
    df1.columns = ["시가", "시가총액"]
    df1 = df1.sort_values('시가총액')
    df1['group'] = pd.cut(df1.reset_index().index, bins=3, labels=['소형주', '중형주', '대형주'])
    
    df2 = stock.get_market_fundamental_by_ticker(f"{year}0101")
    df2 = df2[['PER', 'PBR']]
    
    df3 = stock.get_market_ohlcv_by_ticker(f"{year}1231", prev=True)
    df3 = df3[['종가']]
    
    t0 = pd.merge(left=df1, right=df2, left_index=True, right_index=True)
    df = pd.merge(left=t0, right=df3, left_index=True, right_index=True)
    
    df = df.query('PBR != 0').copy()
    df['수익률'] = df['종가'] / df['시가']
    cond = (df['PER'] >= 2.5) & (df['PER'] <= 10)
    top30 = df[cond].sort_values('PBR').groupby('group').head(30)
    
    how = {
        '수익률' : np.mean
    }
    yoy = top30.groupby('group').agg(how)
    yoy.columns = [year]
    return yoy
```

```python
import time 

dfs = [ ]
for date in range(2010, 2021):
    df = low_per_pbr(f"{date}")
    dfs.append(df)
    time.sleep(1)
    
df = pd.concat(dfs, axis=1)
df
```

```python
# =========================================
# 12.1 저 PER 전략
# 목표: 특정 기준일(2021-04-01) PER이 낮은 종목일수록 수익률(등락률)이 높은지 확인
# =========================================
import pandas as pd
import numpy as np

# 1) 팩터 데이터 불러오기: PER/PBR
df_factor = pd.read_excel(
    "data/data_kosdaq_20210401_per.xlsx",
    index_col=0,
    usecols=[0, 1, 6, 8]   # [종목코드, 종목명, PER, PBR]
)

# display(df_factor)

# TODO: '-' 같은 결측 문자열을 NaN으로 바꾸고, PER/PBR을 숫자형으로 바꿔주세요.
# 힌트: df_factor.replace('-', np.nan, inplace=True); pd.to_numeric(..., errors='coerce')
df_factor.replace('-', np.nan, inplace=True)
df_factor["PER"] = pd.to_numeric(df_factor["PER"], errors='coerce')
df_factor["PBR"] = pd.to_numeric(df_factor["PBR"], errors='coerce')
# 2) 거래량 데이터 붙이기
df_sise = pd.read_excel("data/data_kosdaq_20210401_sise.xlsx", index_col=0)
df_volume = df_sise[['거래량']]  # 필요한 열만 추출
# TODO: PER/PBR 프레임(df_factor)에 거래량(df_volume)을 인덱스 기준으로 조인하세요.
# 힌트: df_factor.join(df_volume)
df2 = df_factor.join(df_volume)

# 3) 등락률(수익률) 데이터 붙이기
df_change = pd.read_excel("data/data_kosdaq_change_2021.xlsx", index_col=0, usecols=[0, 5])  # 등락률
# TODO: df2에 df_change를 조인해서 df3을 만드세요.
df3 = df2.join(df_change)
# display(df3)
# 4) 거래정지 필터: 거래량 0은 제외
# TODO: 거래량이 0이 아닌 종목만 남겨 df4를 만드세요.
# 힌트: cond = df3['거래량'] != 0; df4 = df3[cond].copy()
cond = df3['거래량'] != 0
df4 = df3.loc[cond].copy()
# display(df4)
# 5) PER 기준 정렬(오름차순) 후 하위 30개 선택
# 주의: PER이 NaN/음수/0인 경우는 제외하는 게 안전(선택)
# TODO: df5를 PER 오름차순으로 정렬하고, 인덱스를 리셋하세요.
cond = (df4["PER"].notnull()) & (df4["PER"] > 0)
df5 = df4[cond].sort_values(by="PER").reset_index()
# TODO: 하위 30개를 low_per30으로 저장하고, 평균 등락률을 출력하세요.
low_per30 = df5.head(30)
print("저 PER 하위 30 평균 등락률:", low_per30['등락률'].mean())

# 6) PER 20분위 그룹 나누기 후 그룹별 평균 등락률
# 아이디어: 정렬 순서대로 인덱스를 20개 ‘통’으로 나눠 평균 비교
df5['group'] = pd.cut(df5.index, bins=20, labels=False)
df6 = df5.groupby('group')[['등락률']].mean()
print(df6.head())

# (선택) 시각화
import matplotlib.pyplot as plt
import platform
plt.rc('font', family='Malgun Gothic' if platform.system() != 'Darwin' else 'AppleGothic')
ax = df6['등락률'].plot.bar(figsize=(12,6), title="PER 분위 그룹별 평균 등락률")
ax.set_xlabel("PER 분위 그룹(낮을수록 저PER)")
ax.set_ylabel("평균 등락률")
plt.show()


# =========================================
# 12.2 PBR + PER 콤보 전략
# 목표: PER이 너무 싸지도(이상치) 비싸지도 않은 구간에서 PBR이 낮은(자산가치 대비 저평가) 종목 선별
# =========================================

# TODO: PER이 [2.5, 10] 범위인 종목만 필터해 df5_combo를 만드세요.
# 힌트: cond = (df4['PER'] >= 2.5) & (df4['PER'] <= 10)
cond = (df4['PER'] >= 2.5) & (df4['PER'] <= 10)
df5_combo = df4[cond]

# TODO: df5_combo를 PBR 오름차순으로 정렬 후 상위 30개를 df6_combo에 담으세요.
df6_combo = df5_combo.sort_values(by="PBR").head(30)

# TODO: 그 30개 종목의 기술통계(describe)와 최저 등락률 종목을 확인해보세요.
print(df6_combo.describe())
print(df6_combo[df6_combo['등락률'] == df6_combo['등락률'].min()])


# =========================================
# 12.3 시가총액별 콤보 전략 (학생용)
# 목표: 연초 특성(시가총액, PER/PBR)로 연말 성과(종가/시가)를 예측해보기
# =========================================
from pykrx import stock
import time

# 1) 연도 한 개 예시(2010)
# TODO: 연초 시가총액/시가(단순화 지표)를 불러와 작은 것부터 정렬 후, 3등분 그룹 라벨(소/중/대형주) 부여
df1 = stock.get_market_cap_by_ticker("20100104")[["종가", "시가총액"]].rename(columns={"종가":"시가"})
df1 = df1.sort_values("시가총액")
df1["group"] = pd.cut(df1.reset_index().index, bins=3, labels=["소형주","중형주","대형주"])

# display(df1)
# TODO: 같은 날짜의 펀더멘털(PER/PBR)을 불러오고 필요한 열만 남기세요.
df2 = stock.get_market_fundamental("20100104")
df2 = df2[['PER', 'PBR']]   


# TODO: 연말(2010-12-31 직전 영업일)의 종가를 불러와 df3_end로 저장하세요.
df3_end = stock.get_market_ohlcv("2010-12-31")
df3_end = df3_end['종가']

# TODO: 인덱스 기준 병합(좌→우 순서)으로 t0, df_all을 만들고, PBR=0 제거, '수익률'=연말종가/연초시가 생성
t0 = pd.merge(left=df1, right=df2, left_index=True, right_index=True)
df_all = pd.merge(left=df3_end, right=t0, left_index=True, right_index=True)
df_all = df_all.query("PBR != 0").copy()
df_all["수익률"] = df_all['종가'] / df_all['시가']

# TODO: PER [2.5,10] & PBR 낮은 순 상위 30개를 그룹별(head 30) 뽑아 그룹 평균 수익률 집계
cond = (df_all['PER'] >= 2.5) & (df_all["PER"] <= 10)
top30 = df_all[cond].sort_values("PBR").groupby("group").head(30)
yoy_2010 = top30.groupby("group").agg({"수익률":"mean"})
yoy_2010.columns = ["2010"]
print(yoy_2010)

# 2) 함수화: 연도별 동일 로직 반복
def low_per_pbr(year):
    # TODO: 위 2010 예시를 일반화하여 year 입력만 바꾸면 돌아가도록 작성
    # 힌트: f"{year}0101", f"{year}1231" 포맷 / get_market_ohlcv_by_ticker(..., prev=True)
    date_start = stock.get_nearest_business_day_in_a_week(f"{year}0101")
    df1 = stock.get_market_cap(date_start)
    df1 = df1[["종가", "시가총액"]]
    df1.columns = ["시가", "시가총액"]
    df1 = df1.sort_values('시가총액')
    df1['group'] = pd.cut(df1.reset_index().index, bins=3, labels=['소형주', '중형주', '대형주'])

    df2 = stock.get_market_fundamental(date_start)
    df2 = df2[['PER', "PBR"]]

    date_end = stock.get_nearest_business_day_in_a_week(f"{year}1231")
    df3 = stock.get_market_ohlcv_by_ticker(date_end)
    df3 = df3["종가"]
    t0 = pd.merge(left=df1, right=df2, left_index=True, right_index=True)
    df = pd.merge(left=df3, right=t0, left_index=True, right_index=True)
    
    df = df.query("PBR != 0").copy()
    df["수익률"] = df['종가'] / df['시가']

    cond = (df['PER'] >= 2.5) & (df["PER"] <= 10)
    top30 = df[cond].sort_values("PBR").groupby("group").head(30)

    how = {
        '수익률' : np.mean

    }

    yoy = top30.groupby("group").agg(how)
    yoy.columns = [year]
    
    return yoy
# 3) 2010~2020 반복 실행 (요청 과다 방지 sleep)
dfs = []
for y in range(2010, 2021):
    out = low_per_pbr(y)
    dfs.append(out)
    time.sleep(1)
   
print(dfs)
# 4) 연도별 집계 결합 및 누적 성과 확인
df_years = pd.concat(dfs, axis=1)
print(df_years)

# TODO: 연도별 평균 수익률을 누적으로 곱해(행은 그룹, 열은 연도) 성과 곡선을 그리고, 전치 후 선그래프
cum = df_years.cumprod(axis=1)
display(cum)
cum.T.plot.line(title="시가총액 그룹별 콤보 전략 누적 성과(연도별 곱)")
plt.show()


# =========================================
# 🔍 심화: 실무에 가까운 안전장치(선택)
# =========================================
# 1) 유동성 필터: 거래대금 상위 n%만 대상 (예: 낮은 유동성 종목 제거)
# 2) 이상치 처리: PER/PBR 상하위 1% 절단(winsorize) 또는 quantile 필터 사용
# 3) 그룹 나누기: pd.cut 대신 pd.qcut(분위수 기반)으로 그룹 균형 맞추기
# 4) 수익률 정의: 당일 등락률 대신, T+1~T+n 구간 수익률로 바꾸어 룩어헤드 완화
```

```python
from pykrx.website.krx.market import get_nearest_business_day_in_a_week

def low_per_pbr(year):
    start = f"{year}0101"
    end = get_nearest_business_day_in_a_week(f"{year}1231")

    df1 = stock.get_market_cap(start)[["종가", "시가총액"]]
    df1.columns = ["시가", "시가총액"]
    df1 = df1.sort_values('시가총액')
    df1['group'] = pd.cut(df1.reset_index().index, bins=3, labels=['소형주', '중형주', '대형주'])

    df2 = stock.get_market_fundamental(start)[['PER', "PBR"]]
    df3 = stock.get_market_ohlcv_by_ticker(end, market="KOSPI")[["종가"]]

    t0 = pd.merge(df1, df2, left_index=True, right_index=True, how="inner")
    df = pd.merge(df3, t0, left_index=True, right_index=True, how="inner")

    df = df.query("PBR != 0").copy()
    df["수익률"] = df['종가'] / df['시가']

    cond = (df['PER'] >= 2.5) & (df["PER"] <= 10)
    top30 = df[cond].sort_values("PBR").groupby("group").head(30)

    yoy = top30.groupby("group").agg({'수익률': 'mean'})
    yoy.columns = [year]

    return yoy

dfs = []
for y in range(2010, 2021):
    out = low_per_pbr(y)
    dfs.append(out)
    time.sleep(1)
   
print(dfs)
# 4) 연도별 집계 결합 및 누적 성과 확인
df_years = pd.concat(dfs, axis=1)
print(df_years)

# TODO: 연도별 평균 수익률을 누적으로 곱해(행은 그룹, 열은 연도) 성과 곡선을 그리고, 전치 후 선그래프
cum = df_years.cumprod(axis=1)
display(cum)
cum.T.plot.line(title="시가총액 그룹별 콤보 전략 누적 성과(연도별 곱)")
plt.show()
```
