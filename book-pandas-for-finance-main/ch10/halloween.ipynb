```python
def 투자6개월(df, start_year=2000, end_year=2020, month=11):    
    누적수익률 = 1
    
    for year in range(start_year, end_year): 
        start = datetime.datetime(year=year, month=month, day=1)
        end = start + relativedelta(months=5)
        
        buy_mon = start.strftime("%Y-%m")
        sell_mon = end.strftime("%Y-%m")
        
        # 매수가
        매수가 = df.loc[buy_mon].iloc[0]['Open']
    
        # 매도가
        매도가 = df.loc[sell_mon].iloc[-1]['Close']
    
        # 수익률 
        수익률 = 매도가/매수가
        누적수익률 = 누적수익률 * 수익률
    return 누적수익률
```

```python
# =========================================
# 0) 준비: 데이터 불러오기 (실제 KOSPI)
# =========================================
import numpy as np
import pandas as pd
%pip install -U finance-datareader
plt.rc('font', family="Malgun Gothic")

try:
    import FinanceDataReader as fdr
    kospi = fdr.DataReader(symbol="KS11", start="2000-01-01")  # 실제 KOSPI
    kospi = kospi[["Open","Close"]].copy()
    kospi.index = pd.to_datetime(kospi.index)
    kospi = kospi.sort_index()
    print("FDR로 KOSPI 불러오기 성공:", kospi.shape)
except Exception as e:
    print("⚠️ FDR 불가 → 샘플 KOSPI 생성:", e)
    np.random.seed(42)
    dates = pd.date_range("2000-01-01", periods=7000, freq="B")   # 약 28년 영업일
    close = 1000 * (1 + np.random.normal(0, 0.004, len(dates))).cumprod()
    open_ = np.r_[close[0], close[:-1]]
    kospi = pd.DataFrame({"Open": open_, "Close": close}, index=dates)

# 월 라벨 접근을 위해 인덱스가 DatetimeIndex인지 확인
assert isinstance(kospi.index, pd.DatetimeIndex)

# =========================================
# 10.1 할로윈 전략 (11~다음해 4월)
# =========================================
# TODO 10.1-1) 2000~2020년까지 반복하며 11월 첫 거래일 Open으로 매수, 이듬해 4월 마지막 거래일 Close로 매도
누적수익률 = 1.0
for year in range(2000, 2021):
    buy_mon  = f"{year}-11"      # 11월
    sell_mon = f"{year+1}-04"    # 이듬해 4월
    # 매수가(11월 첫 거래일 Open)
    매수가 = kospi.loc[buy_mon].iloc[0]['Open']
    # 매도가(4월 마지막 거래일 Close)
    매도가 = kospi.loc[sell_mon].iloc[-1]['Close']
    # 기간 수익률
    수익률 = 매도가 / 매수가
    # 누적 곱
    누적수익률 = 누적수익률 * 수익률
print("할로윈 누적수익률:", 누적수익률)

# TODO 10.1-2) CAGR 계산 (투자기간 = 21년)
CAGR = (누적수익률 ** (1/21)) - 1
print("할로윈 CAGR(%)", CAGR * 100)

# TODO 10.1-3) 단순 보유 전략과 CAGR 비교 (기간 동일 21년)
단순보유누적 = kospi.iloc[-1]['Close'] / kospi.iloc[0]['Open']
단순보유CAGR = (단순보유누적 ** (1/ 21)) - 1
print("단순보유 누적/연복리(%)", (단순보유누적-1)*100, 단순보유CAGR*100)

# =========================================
# 10.2 범용 백테스트 함수(6개월 보유)
# =========================================
import datetime
from dateutil.relativedelta import relativedelta

# TODO 10.2-1) 시작월부터 5개월 뒤(총 6개월)까지 보유, 연도 구간 반복 → 누적수익률 반환
def 투자6개월(df, start_year=2000, end_year=2021, month=11):
    누적 = 1.0
    for y in range(start_year, end_year):
        start = datetime.datetime(year=y, month=month, day=1)
        end   = start + relativedelta(months=5)
        buy_mon  = start.strftime("%Y-%m")
        sell_mon = end.strftime("%Y-%m")
        # 매수가(시작월 첫 거래일 Open)
        매수가 = df.loc[buy_mon].iloc[0]['Open']
        # 매도가(끝월 마지막 거래일 Close)
        매도가 = df.loc[sell_mon].iloc[-1]['Close']
        누적 *= (매도가 / 매수가)
    return 누적

# TODO 10.2-2) 1~12월 각각 시작했을 때 6개월 전략 성과 비교
for m in range(1, 13):
    ret = 투자6개월(kospi, 2000, 2021, m)
    print(f"{m:02}월 시작 6개월 누적수익률: {ret:.3f}")

# =========================================
# 10.3 시각화 (막대 그래프)
# =========================================
# TODO 10.3-1) 월별 6개월 전략 누적수익률을 dict로 모아 막대차트 그리기
data = {}
for m in range(1, 13):
    data[m] = 투자6개월(kospi, 2000, 2021, m)

# 막대 차트
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12, 6))
ax  = fig.add_subplot(1, 1, 1)
ax.bar(list(data.keys()), list(data.values()))
ax.set_title("코스피: 시작월별 6개월 전략 누적수익률 (2000~2020)")
ax.set_xlabel("투자 시작월")
ax.set_ylabel("누적수익률")
plt.show()
```
