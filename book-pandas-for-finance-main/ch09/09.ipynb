```python
# =========================================
# 0) 라이브러리 임포트 & 날짜 설정
# =========================================
from pykrx import stock
import pandas as pd

start = "20240101"   # 시작일(YYYYMMDD)
end   = "20240930"   # 종료일(YYYYMMDD)
mk    = "KOSPI"      # "KOSPI" 또는 "KOSDAQ"

# =========================================
# 1) 티커 조회 & 이름 붙이기
# =========================================
# TODO 1-1) 특정 시장(mk)의 티커 리스트를 가져와 tickers에 저장
tickers = stock.get_market_ticker_list(market=mk)


# TODO 1-2) 위에서 가져온 tickers 중 앞에서 5개만 이름과 함께 DataFrame으로 만들기
# 힌트: stock.get_market_ticker_name(ticker)

ticker_name = {}
print("샘플 티커 5개:")
for ticker in tickers[:5]:
    sample = stock.get_market_ticker_name(ticker)
    ticker_name.update({ticker : sample})
    print(ticker, sample)
print(ticker_name)
df = pd.DataFrame(ticker_name.items(), columns=["티커", "종목명"])
display(df)

# =========================================
# 2) OHLCV 조회 (단일 종목)
# =========================================
# TODO 2-1) 아무 티커 하나 고르고(예: sample의 첫 행) 그 티커의 OHLCV를 ohlcv에 저장
# 힌트: stock.get_market_ohlcv_by_date(start, end, ticker, adjusted=True)
ticker = df.iloc[0,0]
ohlcv = stock.get_market_ohlcv_by_date(start, end, ticker, adjusted=True)
print("OHLCV(head):\n", ohlcv.head())

# TODO 2-2) 수익률(일간) 컬럼 'ret' 추가: 종가.pct_change()
ohlcv["ret"] = ohlcv["종가"].pct_change()
print("수익률 통계:\n", ohlcv["ret"].describe())

# =========================================
# 3) 펀더멘털(BPS/PER/PBR/EPS/DIV/DPS) 조회
# =========================================
# TODO 3-1) 위 ticker에 대한 펀더멘털을 fundamental에 저장
# 힌트: stock.get_market_fundamental_by_date(start, end, ticker)
fundamental = stock.get_market_fundamental_by_date(start, end, ticker)
print("펀더멘털(head):\n", fundamental.head())

# TODO 3-2) PER < 15 & PBR < 1.5 인 날짜만 필터링해 value_days에 저장
value_days = fundamental.loc[(fundamental["PER"] < 15) & (fundamental["PBR"] < 1.5)]
print("밸류 조건 일자 수:", len(value_days))

# =========================================
# 4) 멀티 티커 포트폴리오 프레임 만들기
# =========================================
# TODO 4-1) tickers에서 앞 3개를 선택해 멀티 종목 종가만 묶은 DataFrame 'close_wide' 만들기
#  - 각 종목의 종가를 열로, 날짜를 인덱스로 (outer join으로 날짜 정렬)
# 힌트: 루프 + pd.concat(axis=1)
sel = tickers[:3]
close_wide = pd.concat(
    [stock.get_market_ohlcv_by_date(start, end, t, adjusted=True)["종가"].rename(t) 
     for t in sel], axis=1
)
print("멀티 종목 종가(head):\n", close_wide.head())

# TODO 4-2) 각 종목 일간 수익률로 변환해 ret_wide에 저장 (pct_change)
ret_wide = close_wide.pct_change()
print("멀티 종목 수익률(head):\n", ret_wide.head())

# TODO 4-3) 동일가중(각 1/len(sel)) 포트폴리오 수익률 'pf_ret' 시리즈 만들기 (행 평균)
pf_ret = ret_wide.mean(axis=1)
print("포트폴리오 수익률(head):\n", pf_ret.head())

# =========================================
# 5) 응용: 이동평균 & 신호
# =========================================
# TODO 5-1) 단일 종목(위 ticker)의 5일, 20일 이동평균(ma5, ma20) 추가
ohlcv["ma5"] = ohlcv["종가"].rolling(window=5).mean()
ohlcv["ma20"] = ohlcv["종가"].rolling(window=20).mean()

# TODO 5-2) 매수 신호: ma5가 ma20을 위로 돌파하는 날(True), 아니면 False → 'buy' 컬럼
# 힌트: (ma5 > ma20) & (어제 ma5 <= 어제 ma20). shift(1) 활용
ohlcv["buy"] = (ohlcv["ma5"] > ohlcv["ma20"]) & (ohlcv["ma5"].shift(1) <= ohlcv["ma20"].shift(1))
print("매수 신호(head 15):\n", ohlcv[["종가", "ma5", "ma20", "buy"]].head(15))

# TODO 5-3) 신호를 다음날 수익률에 적용하여 'strategy_ret' 계산
# 힌트: ohlcv["ret"].shift(-1) * ohlcv["buy"]
ohlcv["strategy_ret"] = ohlcv["ret"].shift(-1) * ohlcv["buy"]
cum = (1 + ohlcv["strategy_ret"].fillna(0)).cumprod() - 1
print("전략 누적수익률(마지막):", cum.dropna().iloc[-1])
```

```python
from pykrx import stock
import pandas as pd

# =========================================
# 0) 준비
# =========================================
start = "20250101"
end   = "20250630"
market = "KOSPI"

# =========================================
# 1) 종목 선택
# =========================================
tickers = stock.get_market_ticker_list(end, market=market)
sample = pd.DataFrame({
    "코드": tickers[:3],
    "이름": [stock.get_market_ticker_name(t) for t in tickers[:3]]
})
print("샘플 종목 3개:\n", sample)

# 하나만 골라서 사용
ticker = sample["코드"].iloc[0]

# =========================================
# 2) 시세 데이터 가져오기
# =========================================
ohlcv = stock.get_market_ohlcv_by_date(start, end, ticker, adjusted=True)
print("시세 데이터:\n", ohlcv.head())

# 종가 기준으로 변화율(=수익률) 계산
ohlcv["수익률"] = ohlcv["종가"].pct_change()

# =========================================
# 3) 이동평균선 만들기
# =========================================
# 5일, 20일 이동평균 → 단기/중기 흐름 비교
ohlcv["5일평균"] = ohlcv["종가"].rolling(5).mean()
ohlcv["20일평균"] = ohlcv["종가"].rolling(20).mean()

# =========================================
# 4) 매수 신호 만들기
# =========================================
# 🤔 왜 만들까?
# - 가격이 단기간(5일평균)과 장기간(20일평균)을 비교했을 때
#   단기간 평균이 장기간 평균을 "위로 뚫고 올라가는 순간"을 
#   간단한 매수 시점으로 잡아보는 것입니다.
# - 이 조건을 True/False로 기록해두면 나중에 "언제 샀다"를 쉽게 추적할 수 있습니다.
# 📝 어떻게 만들까?
# 1) 오늘: 5일평균 > 20일평균
# 2) 어제: 5일평균 <= 20일평균
# 3) 두 조건이 동시에 맞을 때 → True (매수신호)
ohlcv["매수신호"] = (ohlcv["5일평균"] > ohlcv["20일평균"]) & \
                  (ohlcv["5일평균"].shift(1) <= ohlcv["20일평균"].shift(1))

print("매수 신호 예시:\n", ohlcv[["종가","5일평균","20일평균","매수신호"]].head(30))

# =========================================
# 5) 전략 수익률 계산
# =========================================
# 🤔 왜 만들까?
# - 신호가 True인 날은 "오늘 매수했다"라는 뜻입니다.
# - 하지만 매수 후의 수익은 "다음날부터" 발생합니다.
#   → 따라서 신호가 True인 날에는 다음날 수익률을 전략 수익률로 기록해야 합니다.

# 📝 어떻게 만들까?
# 1) ohlcv["수익률"].shift(-1) → 수익률을 하루 뒤로 당겨서 "다음날 수익률"로 만듭니다.
# 2) 여기에 매수신호(True=1, False=0)를 곱합니다.
#    → 신호가 있으면 다음날 수익률 반영, 신호가 없으면 0.
ohlcv["전략수익률"] = ohlcv["수익률"].shift(-1) * ohlcv["매수신호"].astype(float)

# 누적 수익률 계산
ohlcv["누적수익률"] = (1 + ohlcv["전략수익률"].fillna(0)).cumprod() - 1

print("누적 수익률 마지막 값:", ohlcv["누적수익률"].dropna().iloc[-1])
```

```python
from pykrx import stock
import pandas as pd
import numpy as np

# =========================================
# 0) 준비
# =========================================
start = "20250101"   # 시작 날짜 (예: 2025년 1월 1일)
end   = "20250630"   # 끝 날짜   (예: 2025년 6월 30일)
market = "KOSPI"     # 시장 종류 ("KOSPI" 또는 "KOSDAQ")
short_n = 5          # 짧은 이동평균(며칠 평균?)
long_n  = 20         # 긴   이동평균(며칠 평균?)
fee_bp  = 0.0005     # 거래 비용 비율(예: 0.05% → 0.0005)

# =========================================
# 1) 종목 고르기 (3개만 사용)
# =========================================
tickers = stock.get_market_ticker_list(end, market=market)
# TODO 1-1) 위 리스트에서 앞 3개만 골라 sel 변수에 담아보세요.
sel = tickers[:3]

# (보조) 코드 → 종목명 매핑
names = {t: stock.get_market_ticker_name(t) for t in sel}
print("선택 종목:", {t: names[t] for t in sel})

# =========================================
# 2) 전략 함수 만들기
# =========================================
def make_signal_df(ticker, start, end, short_n=5, long_n=20, fee_bp=0.0,
                   use_stop=True, stop_loss=-0.05, take_profit=0.10):
    df = stock.get_market_ohlcv_by_date(start, end, ticker, adjusted=True).copy()
    if df.empty or "종가" not in df.columns:
        raise ValueError(f"{ticker}: 데이터가 없거나 '종가'가 없습니다.")
    
    # 2-1) 기본 계산
    df["ret"]  = df["종가"].pct_change()
    # TODO 2-1) 짧은선/긴선 이동평균을 계산해 'ma_s', 'ma_l' 컬럼에 저장하세요.
    df["ma_s"] = df["종가"].rolling(window=short_n).mean()
    df["ma_l"] = df["종가"].rolling(window=long_n).mean()
    
    # 2-2) 매수 신호
    cross_today = df["ma_s"] > df["ma_l"]
    cross_yday  = df["ma_s"].shift(1) <= df["ma_l"].shift(1)
    df["buy_sig"] = cross_today & cross_yday
    
    # 2-3) 보유 상태 만들기
    
    df["pos"] = 0
    holding = 0
    entry_price = np.nan
    
    cross_down_today = df["ma_s"] <= df["ma_l"]
    cross_down_yday  = df["ma_s"].shift(1) >  df["ma_l"].shift(1)
    
    for i in range(len(df)):
        if holding == 0 and df["buy_sig"].iloc[i]:
            holding = 1
            entry_price = df["종가"].iloc[i]
        elif holding == 1 and not np.isnan(entry_price):
            if use_stop:
                pnl = (df["종가"].iloc[i] / entry_price) - 1
                # TODO 2-3) 손절/익절 조건(pnl <= stop_loss 또는 pnl >= take_profit)이면 보유 종료하세요.
                if pnl <= stop_loss or pnl >= take_profit:
                    holding = 0
                    entry_price = np.nan
        if holding == 1 and cross_down_today.iloc[i] and cross_down_yday.iloc[i]:
            holding = 0
            entry_price = np.nan
        df.at[df.index[i], "pos"] = holding
    
    # 2-4) 전략 수익률
    # 🤔 룩어헤드(look-ahead bias)란?
    #   - "내가 미래를 미리 알고 매매한 것처럼 계산하는 오류"를 말합니다.
    #   - 예: 오늘 신호가 나왔는데, 오늘 종가를 기준으로 수익을 기록하면
    #         실제 매매에서는 불가능한 '미래 정보'를 사용한 꼴이 됩니다.
    #   - 따라서 "신호가 나온 다음 날"부터 수익을 기록해야 현실적입니다.
    raw_strategy = df["ret"].shift(-1) * df["pos"]
    
    # 2-5) 거래비용 반영
    trade = df["pos"].diff().abs().fillna(0)
    # TODO 2-5) 비용 차감 반영: strategy = raw_strategy - trade * fee_bp
    strategy = raw_strategy - trade * fee_bp
    
    df["strategy_ret"] = strategy
    df["cum"] = (1 + df["strategy_ret"].fillna(0)).cumprod() - 1
    return df

# =========================================
# 3) 3개 종목 실행
# =========================================
# TODO 3-1) sel의 각 종목을 make_signal_df로 처리하여 res 딕셔너리에 저장하세요.
# 힌트: {티커: 데이터프레임, ...}
res = {t: make_signal_df(t, start, end, short_n, long_n, fee_bp) for t in sel}

# =========================================
# 4) 거래 기록(체결 로그)
# =========================================
def make_tradelog(df, ticker):
    log = []
    pos_prev = 0
    for i in range(len(df)):
        pos_now = int(df["pos"].iloc[i])
        if pos_prev == 0 and pos_now == 1:
            log.append({"ticker": ticker, "side": "BUY", "date": df.index[i], "price": df["종가"].iloc[i]})
        elif pos_prev == 1 and pos_now == 0:
            log.append({"ticker": ticker, "side": "SELL", "date": df.index[i], "price": df["종가"].iloc[i]})
        pos_prev = pos_now
    return pd.DataFrame(log)

# 4-1) 세 종목의 로그 합치기
# 🤔 왜 합칠까?
#   - 개별 종목마다 매매 기록을 따로 두면 전체 거래 흐름을 한눈에 보기 어렵습니다.
#   - 따라서 모든 종목의 거래 기록을 하나로 합쳐 "언제 어떤 종목을 사고팔았는지"를
#     시간순으로 쉽게 확인할 수 있게 합니다.
# TODO 4-1) 세 종목의 로그를 합쳐 logs로 만들고 출력하세요.
# 힌트: pd.concat([...], ignore_index=True)
logs = pd.concat([make_tradelog(res[t], t) for t in sel], ignore_index=True)
print("\n=== 체결 로그(샘플) ===")
print(logs.head(10))

# =========================================
# 5) 종목별 성과 요약
# =========================================
def compute_winrate(log):
    win, tot = 0, 0
    entry = None
    for _, r in log.iterrows():
        if r["side"] == "BUY":
            entry = r["price"]
        elif r["side"] == "SELL" and entry is not None:
            tot += 1
            if r["price"] > entry:
                win += 1
            entry = None
    return (win / tot) if tot > 0 else np.nan

summary = []
for t in sel:
    df = res[t]
    # TODO 5-1) 마지막 누적 수익률 구하기 (빈 경우 0.0)
    last_cum =  df["cum"].dropna().iloc[-1] if not df["cum"].dropna().empty else 0.0
    # TODO 5-2) 승률 계산하기: compute_winrate 호출
    wr = compute_winrate(make_tradelog(df, t))
    summary.append({"ticker": t, "name": names[t], "last_cum": last_cum, "winrate": wr})

summary_df = pd.DataFrame(summary).sort_values("last_cum", ascending=False)
print("\n=== 종목별 성과 비교 ===")
print(summary_df)
```
