```python
import pandas as pd

# ---------------------------------------------------------
# 초기 데이터
# ---------------------------------------------------------
data = [
    ["사과", 1000, 50],
    ["바나나", 2000, 80],
    ["체리", 1500, 30]
]
index = ["F001", "F002", "F003"]
columns = ["상품명", "가격", "재고"]
df = pd.DataFrame(data=data, index=index, columns=columns)
df.index.name = "상품코드"
print("원본 DataFrame:\n", df, "\n")

# ---------------------------------------------------------
# 1) 할인율 컬럼 추가 & 신상품 행 추가
# ---------------------------------------------------------
# TODO-1: 새 컬럼 '할인율' 추가 (예: [0.10, 0.20, 0.30])
df["할인율"] = [0.10, 0.20, 0.30]

# TODO-2: 신상품 'F004' 추가 (예: ["포도", 3000, 20, 0.15])
df.loc["F004"] = ["포도", 3000, 20, 0.15]

print("[1] 할인율 추가 & 신상품 추가:\n", df, "\n")

# ---------------------------------------------------------
# 2) 파생 컬럼 계산
# ---------------------------------------------------------
# TODO-3: '할인가격' = 가격 * (1 - 할인율)
df["할인가격"] = df['가격'] * (1 - df['할인율'])

# TODO-4: '매출가능액(할인전)' = 가격 * 재고
df["매출가능액(할인전)"] = df['가격'] * df['재고']

# TODO-5: '매출가능액(할인후)' = 할인가격 * 재고
df["매출가능액(할인후)"] = df['할인가격'] * df['재고']

print("[2] 파생 컬럼 추가:\n", df, "\n")

# ---------------------------------------------------------
# 3) 필터링 & 정렬
# ---------------------------------------------------------
# TODO-6: 재고가 40 이상이고, 할인가격이 1500 이하인 상품만 추출 -> sub
sub = df[(df['재고'] >= 40) & (df['할인가격'] <= 1500) ]
print("[3] 필터링 결과(sub):\n", sub, "\n")

# TODO-7: 매출가능액(할인후) 기준 내림차순 정렬 -> ranked
ranked = df.sort_values(by='매출가능액(할인후)', ascending=False)
display("[3] 매출가능액(할인후) 내림차순 정렬:\n", ranked[["상품명","할인율","할인가격","재고","매출가능액(할인후)"]], "\n")

# ---------------------------------------------------------
# 4) 최고 매출가능액 상품 찾기 (idxmax)
# ---------------------------------------------------------
# TODO-8: '매출가능액(할인후)'가 가장 큰 상품 코드와 값
best_code = ranked['매출가능액(할인후)'].idxmax()
best_value = ranked.loc[best_code, '매출가능액(할인후)']
print(f"[4] 최고 매출가능액 상품: {best_code} (금액 {int(best_value)}원)\n")

# ---------------------------------------------------------
# 5) 리네이밍 (컬럼/인덱스)
# ---------------------------------------------------------
# TODO-9: 컬럼명을 ['name','price','stock','discount','price_after','rev_before','rev_after'] 로 변경
#         인덱스명을 'code'로 변경
ranked.columns = ['name','price','stock','discount','price_after','rev_before','rev_after']
ranked.index.name = "code"
display("[5] 리네이밍 결과:\n", ranked, "\n")
```

```python
import pandas as pd
import numpy as np

# --- 예시 데이터 ---
np.random.seed(0)
days = pd.date_range("2025-08-01", periods=7, freq="D")

# (A) 상품 마스터: 코드, 이름, 카테고리, 가격
master = pd.DataFrame({
    "code": ["F001","F002","F003","F004","F005"],
    "name": ["사과","바나나","체리","포도","망고"],
    "category": ["국산","수입","국산","수입","수입"],
    "price": [1000, 2000, 1500, 3000, 3500]
})

# (B) 일자별 판매(개수) 로그: day×code
sales_log = pd.DataFrame({
    "date": np.repeat(days, len(master)),
    "code": list(master["code"]) * len(days),
    "qty": np.random.randint(0, 21, size=len(days)*len(master))
})
display(master)

display(sales_log)
# 1) 판매 로그와 상품 마스터 결합(라벨 붙이기)
# TODO: left merge로 sales_log에 name/category/price 붙이기 → df
df = pd.merge(sales_log, master, on="code", how="inner")

# 2) 일별·상품별 매출액 계산: revenue = price * qty
# TODO: df["revenue"] 생성
df["revenue"] = df["price"] * df["qty"]
display(df)

# 3) 카테고리별 주간 합계 리포트 (qty, revenue)
# TODO: groupby("category")로 qty/revenue 합계 → cat_week
cat_week = df.groupby("category")[["qty", "revenue"]].sum()
display(cat_week)
# 4) 피벗테이블: 행=date, 열=category, 값=revenue (일자×카테고리 매출표)
# TODO: pivot_table로 피벗 생성 → pvt
pvt = pd.pivot_table(data=df, values="revenue", index="date", columns="category")
display(pvt)
# 5) 주간 매출 TOP 카테고리/상품 찾기
# TODO: cat_week에서 revenue 기준 내림차순 정렬 → top_cat
# TODO: 상품(name) 기준으로도 groupby 합계 후 revenue TOP 3 추출 → top_items
top_cat = cat_week.sort_values(by="revenue", ascending=False)
top_items = df.groupby("name")["revenue"].sum().sort_values(ascending=False)[0:3]
print(top_items)

print("=== 결합 DF 샘플 ===\n", df.head(), "\n")
print("=== 카테고리 주간 합계 ===\n", cat_week, "\n")
print("=== 일자×카테고리 피벗(매출) ===\n", pvt, "\n")
print("TOP 카테고리:\n", top_cat, "\n")
print("TOP 3 상품:\n", top_items, "\n")
```

```python
import pandas as pd
import numpy as np

np.random.seed(2)
days = pd.date_range("2025-08-01", periods=7, freq="D")

# 마스터
master = pd.DataFrame({
    "code": ["F001","F002","F003","F004"],
    "name": ["사과","바나나","체리","포도"],
    "price": [1000, 2000, 1500, 3000]
})

# 하루 판매수량(일자×상품)
qty = pd.DataFrame({
    "date": np.repeat(days, len(master)),
    "code": list(master["code"]) * len(days),
    "qty": np.random.randint(0, 31, size=len(days)*len(master))
})

# 결합
df = qty.merge(master, on="code", how="left")

# 시나리오 A: 전상품 10% 할인
# 시나리오 B: 일매출이 낮은 날(총수량<=60)만 20% 할인, 그 외는 0%
# 1) 일자별 총수량 계산해서 df에 붙이기

by_day_qty = df.groupby("date")["qty"].sum()
by_day_qty.rename("day_qty", inplace=True)

df = pd.merge(df, by_day_qty, on="date", how="inner")
df.columns = ["date", "code", "qty", "name", "price", "qty_total"]

# 2) A/B 할인가격 컬럼 만들기(price_A, price_B)
# TODO: price_A = price*0.9
# TODO: price_B = np.where(by_day_qty_total_condition, price*0.8, price)
df["price_A"] = df["price"] * 0.9
df["price_B"] = np.where(df["qty_total"] <= 60, df["price"] * 0.8, df["price"])

# 3) A/B 매출(revenue_A/B) 계산 후, 일자/전체 합 비교
df["revenue_A"] = df["qty"] * df["price_A"]
df["revenue_B"] = df["qty"] * df["price_B"]

daily = df.groupby("date")[["revenue_A", "revenue_B"]].sum()

totalA = df["revenue_A"].sum()
totalB = df["revenue_B"].sum()

# 4) A vs B 차이 큰 상품 Top-3 (전체 합 기준)
diff_by_item = df.groupby(['code', 'name'])[["revenue_A", "revenue_B"]].sum()

diff_by_item['diff'] = abs(diff_by_item["revenue_A"] - diff_by_item["revenue_B"])
top3 = diff_by_item.sort_values(by='diff', ascending=False).head(3)

print("=== 일자별 총수량 ===\n", by_day_qty, "\n")
print("=== 시나리오별 일자 매출 ===\n", daily, "\n")
print("총매출 A:", int(totalA), " / B:", int(totalB))
print("A-B 차이 Top3 상품:\n", top3, "\n")
```
