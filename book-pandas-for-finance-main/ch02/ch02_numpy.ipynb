```python
# 어떤 주식의 5일간 종가
# 하루 수익률 (오늘 가격 - 어제 가격) / 어제 가격
# 종가, 하루 수익률, 평균 수익률, 변동성(표준편차)
price = [100, 102, 98, 105, 110]
yesterday_price = [100]
day_profitrate = []
profit_sum = 0
for i in range(4):
    yesterday_price.append(price[i])
for i in range(5):
    print(f"{i+1}일째 종가는 {price[i]}이다.")
    print(f"{i+1}일째 수익률은 {round((price[i]-yesterday_price[i]) / yesterday_price[i] * 100, 2)}%이다.")
    day_profitrate.append(round((price[i]-yesterday_price[i]) / yesterday_price[i] * 100, 2))
    profit_sum += round((price[i]-yesterday_price[i]) / yesterday_price[i] * 100, 2)
print(f"평균 수익률은 {profit_sum / len(price)}%이다.")
print(round(np.average(day_profitrate),2))
print(f"변동성은 {np.std(day_profitrate)}이다.")
```

```python

scores = np.array([
    [90, 80, 85],
    [70, 75, 80],
    [60, 65, 70]
])

score_mean = np.mean(scores)  #  평균
print(score_mean)
score_std = np.std(scores)   #  표준편차

broadcasted = (scores - score_mean) / score_std
np.mean(broadcasted, axis=1)
np.std(broadcasted, axis=0)
```

```python

import pandas as pd

s = pd.Series([5, 10, 15, 20, 8])


# 1) 조건: 값이 10보다 크면 10을 더하고, 아니면 -10을 빼기
# TODO: np.where 또는 map 을 활용해 구현해보세요.
result = np.where(s > 10, s + 10, s - 10)
print(result)

# 2) 조건: 값이 짝수면 그대로 두고, 홀수면 0으로 바꾸기
# TODO: 조건 연산으로 만들어보세요.
even_only = np.where( s % 2 == 0, s, 0)
print(even_only)

# 3) 조건: 평균보다 크면 "HIGH", 작으면 "LOW"라는 문자열로 바꾸기
# TODO: map 또는 list comprehension을 활용해보세요.
labels = ["HIGH" if x > np.mean(s) else "LOW" for x in s]
print(labels)
```
