```python
import datetime
date = input("검색할 날짜(yyyy-mm-dd):")
if date == "":
    date = datetime.datetime.now().strftime('%Y-%m-%d')
date
```

```python
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
import time
import pandas as pd
driver = Chrome()

page = 1
article_list = []
while True:
    url = f"https://finance.naver.com/news/mainnews.naver?date={date}&page={page}"
    driver.get(url)

    block = driver.find_elements(By.CLASS_NAME, "block1")

    for element in block:
        subject = element.find_element(By.CSS_SELECTOR, ".articleSubject").text
        summary_list = element.find_element(By.CSS_SELECTOR, '.articleSummary').text.split(" ")[:-4]
        summary = " ".join(summary_list)
        press = element.find_element(By.CSS_SELECTOR, '.press').text
        wdate = element.find_element(By.CSS_SELECTOR, '.wdate').text
        link = element.find_element(By.CSS_SELECTOR, ".articleSubject > a").get_attribute("href")
        article_list.append([subject, summary, press, wdate, link]) 

   
    # driver.find_element(By.CSS_SELECTOR, ".pgRR") #none --> False
    try:
        driver.find_element(By.CSS_SELECTOR, ".pgRR")
        page += 1
    except:   
        break

df = pd.DataFrame(article_list, columns=["subject", "summary", "press", "wdate", "link"])
df['date'] = df['wdate'].str[:10]
df = df[['date', "subject", "summary", "press", "wdate", "link"]]
df
```

```python
block = driver.find_elements(By.CLASS_NAME, "block1")
article_list = []
for element in block:
    subject = element.find_element(By.CSS_SELECTOR, ".articleSubject").text
    summary_list = element.find_element(By.CSS_SELECTOR, '.articleSummary').text.split(" ")[:-4]
    summary = " ".join(summary_list)
    press = element.find_element(By.CSS_SELECTOR, '.press').text
    wdate = element.find_element(By.CSS_SELECTOR, '.wdate').text
    link = element.find_element(By.CSS_SELECTOR, ".articleSubject > a").get_attribute("href")
    article_list.append([subject, summary, press, wdate, link])

article_list  
```

```python
import re
from selenium.webdriver.common.by import By
articles = driver.find_elements(By.CSS_SELECTOR, '.block1')
for article in articles:
    summary = article.find_element(By.CSS_SELECTOR, '.articleSummary').text
    summary = summary.replace('\n', ' ').strip()
    
    # 1. 날짜 제거: " | 2025-08-21 11:03:07" 같은 부분 제거
    summary = re.sub(r'\s*\|\s*\d{4}-\d{2}-\d{2}.*$', '', summary)
    
    # 2. 뉴스사 이름 제거: 마지막에 붙은 뉴스사
    summary = re.sub(r'\s+[가-힣A-Za-z·\s]+$', '', summary)
    
    print(summary)
```

```python
import pandas as pd
df = pd.DataFrame(article_list, columns=["subject", "summary", "press", "wdate", "link"])
df['date'] = df['wdate'].str[:10]
df = df[['date', "subject", "summary", "press", "wdate", "link"]]
df
```

```python
from sqlalchemy import create_engine

# -----------------------------------------------------------
# db연결 정보
# db종류+db연결드라이버://username:password@host:port/dbname"
# -----------------------------------------------------------
db_url = "postgresql+psycopg2://postgres:1111@localhost:5432/kiwoom"

# -----------------------------------------------------------
# db연결 엔진 생성
# postgresql+psycopg2://username:password@host:port/dbname"
# -----------------------------------------------------------
engine = create_engine(db_url)
engine 
```

```python
from sqlalchemy.types import Date,TIMESTAMP
from sqlalchemy import text

query = text(f"delete from finance_news where date='{date}';")
with engine.connect() as conn :
    conn.execute(query)
    conn.commit()

df.to_sql('finance_news', engine, if_exists='append', index=False, dtype={"date": Date(), "wdate": TIMESTAMP()})
```
