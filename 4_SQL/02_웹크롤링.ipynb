```python
driver = Chrome()
driver.get("https://www.google.com")

# 'a' 태그(링크) 전부 가져오기
links = driver.find_elements(By.TAG_NAME, "a")

print(f"링크 개수: {len(links)}")
for link in links[:5]:   # 앞에 5개만 출력
    print(link.text)
```
