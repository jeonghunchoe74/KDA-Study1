```python
import pandas as pd
```

### 기상 데이터프레임 생성

* 여러 데이터파일을 연결하여 하나의 데이터프레임 생성
* 데이터 파일 리스트 생성
    * `import glob`
    * `glob.glob(파일경로규칙)`

```python
# ---------------------------
# 데이터 파일 리스트 생성
# ---------------------------

import glob

# data/폴더에서 '기상데이터'가 파일명에 포함된 모든 csv 파일
# 파일명 앞에는 한글자 이상의 다른 글자가 포함되어야 함

files = glob.glob("data/기상데이터/*기상데이터.csv")
files
```

```python
# ---------------------------
# 데이터프레임 리스트 생성
# ---------------------------
temp = pd.read_csv('data/기상데이터/1907-1909 기상데이터.csv', encoding='cp949', parse_dates=['일시'])
df_list = []
for file in files:
    df = pd.read_csv(file, encoding='cp949', parse_dates=['일시'])
    df_list.append(df)
```

```python
# ---------------------------
# 데이터프레임 리스트의 길이 확인
# ---------------------------
len(df_list)


# ---------------------------
# 데이터프레임 리스트의 개별 데이터프레임 확인
# ---------------------------
df_list[4]
```

```python
# ---------------------------
# 데이터프레임 연결
#   df_all
# ---------------------------
df_all = pd.concat(df_list, ignore_index=True)
```

```python
# ---------------------------
# 연결한 데이터프레임 shape 확인
# ---------------------------

df_all.shape

# ---------------------------
# 연결한 데이터프레임 head, tail 확인
# ---------------------------
df_all.head()
df_all.tail()
```

```python
# ---------------------------
# info()
# ---------------------------
df_all.info()
```

```python
# ---------------------------
# 결측치 확인
# ---------------------------
df_all.isnull().sum()
```

```python
# ---------------------------
# 연도, 월 컬럼 추가
# ---------------------------
df_all['연도'] = df_all['일시'].dt.year
df_all['월'] = df_all['일시'].dt.month
```

```python
df_all.head()
```

### 지점정보 데이터프레임 생성
* 지점, 지점명, 데이터수, 시작년도, 끝년도

```python
# --------------------
# 지점별 데이터 수
#   df_location_cnt
# --------------------

# df_location_cnt = df_all.value_counts('지점명')
df_all.groupby('지점명').size().sort_values(ascending=False)
```

```python
# --------------------
# 지점별 데이터 수
#   df_location_cnt
# --------------------

df_location_cnt = df_all.value_counts(['지점','지점명']).to_frame()
df_location_cnt.columns = ["데이터수"]
df_location_cnt
```

```python
# -------------------------
# 지점별 시작년도, 끝년도
#   df_location_year
# -------------------------
df_location_year = df_all.groupby(['지점', '지점명']).agg({'연도':['min', 'max']})
df_location_year.columns = ['시작년도', '끝년도']
df_location_year
```

```python
# -------------------------
# 지점별 데이터 수, 시작년도, 끝년도 concat
#   df_location
# -------------------------
df_location = pd.concat([df_location_cnt, df_location_year], axis=1).reset_index()
df_location
```

### 데이터프레임을 파일로 다운로드

#### 기상데이터

```python
# --------------------
# 파일로 다운로드
#   기상데이터_전체.csv
# --------------------

file_path = 'data/기상데이터/기상데이터_전체.csv'
df_all.to_csv(file_path, index=False)
```

```python
# --------------------
# 파일을 다시 열어서 확인
# --------------------
pd.read_csv(file_path)
```

#### 지점정보

```python
# --------------------
# 파일로 다운로드
#   기상데이터_지점정보.csv
# --------------------
file_path = 'data/기상데이터/기상데이터_지점정보.csv'
df_location.to_csv(file_path, index=False)
```
