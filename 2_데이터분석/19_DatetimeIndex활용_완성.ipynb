## Resampling
* 다양한  주기로 데이터를 집계(aggregate)
    * 더 큰 시간 단위로 데이터를 요약
        * 시간대별 -> 일별, 월별, 연도별
* Offset aliases
    * https://pandas.pydata.org/docs/user_guide/timeseries.html#timeseries-offset-aliases

```python
# -----------------------
# 일 평균
# -----------------------

df_daily = df[['count']].resample(rule="D").mean()
df_daily.head()
```

```python
# -----------------------
# 월 평균(MS)
# -----------------------

df_monthly = df[['count']].resample(rule="MS").mean()
df_monthly.head()
```

```python
df_monthly.plot(kind='bar', 
                title='월 평균 대여량')

xtic = df_monthly.index.strftime('%Y-%m')
plt.xticks(range(len(df_monthly.index)), xtic);
```

```python
# -----------------------
# 연 평균(YS)
# -----------------------

df_yearly = df[['count']].resample(rule="YS").mean()
df_yearly.head()
```

```python
df_yearly.plot(kind='bar',
               figsize=(4,3),
               title='연 평균 대여량')

xtic = df_yearly.index.strftime('%Y')
plt.xticks(range(len(df_yearly.index)), xtic);
```

## Shift
- 시계열 분석 알고리즘을 사용하기 위해 데이터를 임의의 시간만큰 앞 또는 뒤로 이동
- 변화량 계산 등에 활용

```python
# ---------------------------
# 한칸씩 뒤로 이동
# ---------------------------

df['count-1'] = df[['count']].shift(1)
df[['count', 'count-1']].head()
```

```python
# --------------------------------------
# 이전 데이터와의 차이 계산
# --------------------------------------

df['diff'] = df['count'] - df['count-1']
df[['count', 'count-1', 'diff']].head()
```

```python
# ----------------------------------------------------------------
# 미션 
# 1. 일자별로 데이터를 요약한다.
# 2. 이전 일자 대비 증감 count를 구한다.
# 3. 2011년 10월 10~20일 증감 count를 막대그래프로 시각화 한다.
# ----------------------------------------------------------------

# 1
df_daily = df[['count']].resample(rule='D').sum() 

# 2
df_daily['count-1'] = df_daily['count'].shift(1) 
df_daily['diff'] = df_daily['count'] - df_daily['count-1'] 

# 3
df_plot = df_daily.loc['2011-10-10':'2011-10-20','diff']
df_plot.plot(kind='bar')
xtic = df_plot.index.strftime('%Y-%m-%d')
plt.xticks(range(len(df_plot)), xtic);
```

```python
# ----------------------------------------------------------------
# 미션 
# 1. 일자별로 데이터를 요약한다.
# 2. 이전 일자 대비 증감 count를 구한다.
# 3. 2011년 10월 10~20일 증감 count를 막대그래프로 시각화 한다.
# ----------------------------------------------------------------

# 1
df_daily = df[['count']].resample(rule='D').sum() 

# 2
df_daily['count-1'] = df_daily['count'].shift(1) 
df_daily['diff'] = df_daily['count'] - df_daily['count-1'] 

# 3
df_plot = df_daily.loc['2011-10-10':'2011-10-20','diff']
colors = ['red' if val<0 else 'blue' for val in df_plot]
df_plot.plot(kind='bar', color=colors, title='일별 대여량 증감 현황')
xtic = df_plot.index.strftime('%Y-%m-%d')
plt.xticks(range(len(df_plot)), xtic)
plt.axhline(y=0, color='black', linestyle=':')
plt.ylabel('증감 count')
pass
```

## Rolling 
* 이동윈도우
* 고정된 길이의 창(rolling window)을 이동시키면서 각 윈도우 내의 데이터 집계
    * 이동평균, 이동합계 등
    * 데이터 평활화(smoothing)를 통해 노이즈를 제거하고, 단기 추세를 파악할 수 있다
   
* `컬럼.rolling(윈도우크기).통계함수()`

```python
# -------------------
#  샘플 데이터
# -------------------

data = [1,2,3,4,5,6, 7]
idx = pd.date_range('2020-01-01', periods=7, freq='D')
cols = ['col']

df = pd.DataFrame(data=data, index=idx, columns=cols)
display(df)
```

```python
# ---------------------------
# 3일 이동평균
# ---------------------------

df.rolling(window=3).mean()
```

```python
# ---------------------------
# 3일 이동합계
# ---------------------------

df.rolling(window=3).sum()
```

## Expanding
* 확장윈도우
* 시작점부터 데이터를 누적하여 통계 계산
* `컬럼.expanding().통계함수()`

```python
# ---------------------------
# 누적평균
# ---------------------------

df.expanding().mean()
```

```python
# ---------------------------
# 누적 합계
# ---------------------------

df.expanding().sum()
```
