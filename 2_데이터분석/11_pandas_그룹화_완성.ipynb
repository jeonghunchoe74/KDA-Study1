```python
import pandas as pd
import numpy as np
```

```python
# --------------------------
# 샘플데이터
# --------------------------
df = pd.read_csv('data/scores.csv').fillna(0)
df = df.astype({'국어':'int', '수학':'int'})
df['평균'] = df[['국어','영어','수학']].mean(axis=1).round(2)
df
```

### 그룹화 
* `데이터프레임객체.groupby(기준)` --> 그룹객체 리턴
* 데이터를 하나 이상의 열을 기준으로그룹화
* 각 그룹별로 통계 계산(합계, 평균, 개수 등) 수행

```python
# --------------------------
# 성별로 그룹화
# --------------------------

df_group = df.groupby('성별')
df_group
```

### 그룹 객체의 주요 속성
* `그룹객체.groups` : 그룹별 데이터의 인덱스리스트
* `그룹객체.ngroups` : 그룹의 총 개수

```python
# --------------------------
# 그룹별 인덱스리스트
# --------------------------

df_group.groups
```

```python
# --------------------------
# 그룹의 갯수
# --------------------------
df_group.ngroups
```

### 그룹 객체의 주요 메서드
* `그룹객체.get_group()` : 특정 그룹의 데이터 출력
* `그룹객체.describe()` : 그룹별 요약통계

#### get_group()
* `그룹객체.get_group('그룹')`

```python
# --------------------------
# 'f'그룹 데이터 
# --------------------------

df_group.get_group('f')
```

#### 그룹별 요약통계
* `그룹객체.describe()`

```python
# --------------------------
# 그룹별 요약통계
# --------------------------

df_group.describe()
```

#### 그룹별 특정 컬럼의 요약통계
*  `그룹객체[컬럼].describe()`

```python
# --------------------------
# '평균' 컬럼의 요약통계
# --------------------------

df_group['평균'].describe()
```

#### 특정 컬럼의 그룹별 통계값 
* `그룹객체['컬럼명'].통계함수`
    * `count()` : 누락값을 제외한 데이터 수
    * `size()` : 누락값을 포함한 데이터 수
    * `mean()` : 평균
    * `sum()` : 합계
    * `std()` : 표준편차
    * `min()` : 최소값
    * `max()` : 최대값
    * `sum()` : 전체 합

```python
# --------------------------
# 그룹별 총 평균(평균의 평균)
# --------------------------

df_group['평균'].mean()
```

### 특정 컬럼의 여러 통계값 출력
* `그룹객체.agg({컬럼명:통계함수리스트})`

```python
# --------------------------
# 평균 컬럼의 'sum', 'count', 'mean' 
# --------------------------

df_group.agg({'평균':['sum', 'count', 'mean' ]})
```

# 타이타닉 데이터 분석

```python
# --------------------------
# 샘플 데이터
# --------------------------
import seaborn as sns

# 타이타닉 데이터 로드
df_titanic = sns.load_dataset("titanic")
df_titanic = df_titanic.loc[:,['class','who','alone', 'age','survived']].dropna()

# 상위 5개 행 출력
print('shape >>> ', df_titanic.shape)
print('head >>> ')
display(df_titanic.head())
print('info >>> ')
df_titanic.info()
```

#### class별 평균나이

```python
df_titanic.groupby('class')['age'].mean()
```

#### class별 승선자 수

```python
df_titanic.groupby('class')['survived'].count()
```

#### class별 생존자 수

```python
df_titanic.groupby('class')['survived'].sum()
```

#### class별 생존율

```python
df_titanic.groupby('class')['survived'].mean()
```

#### class별 승선자수, 생존자수, 생존율

```python
df_titanic.groupby('class')['survived'].agg(['count','sum','mean'])
```
