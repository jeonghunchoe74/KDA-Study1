## 그래프 설정
* 모든 유형의 그래프에 공통되는 설정
    * `plt.figure()` : 그래프 전체 영역 설정
    * `plt.grid()` : 격자선 표시
    * `plt.title()` : 그래프 제목 설정
    * `plt.legend()` : 범례 설정
    * `plt.xlabel()`, `plt.ylabel()` : x, y축 이름 설정
    * `plt.xlim()`, `plt.ylim()` : 축의 범위 설정
    * `plt.xticks()`, `plt.yticks()` : 축의 눈금 설정

```python
# --------------------------------------------
# 샘플데이터 : 어떤 아이의 나이별 키의 변화
# --------------------------------------------
age = [7,8,9,10,11,12,13,14,15]
height = [122.1, 127.9, 133.4, 138.8, 142.7, 151.4, 155.6, 165.0, 169.2]

# --------------------------------------------
# 그래프 작성 및 설정
# --------------------------------------------

plt.figure(figsize=(4,3))

plt.plot(age, height, 
         color="red",
         linestyle=":",
         marker = ".",
         label = "김철수")
plt.title("oo이의 나이에 따른 키 성장")
plt.legend()
plt.xlabel("age")
plt.ylabel("height(cm)")
plt.ylim(0, 200)
plt.xlim(0, 20)
plt.xticks(range(0,21))
plt.grid()
pass


```

## 막대그래프
* 수치의 크기를 비교할 때 주로 사용
* **`plt.bar(x축데이터,y축데이터)`**
>* 주요옵션
>   * width : 막대의 너비(0~1)

* **`plt.barh(x축데이터,y축데이터)`**
>* 주요옵션
>   * height : 막대의 높이(0~1)

```python
# --------------------------------------------
# 샘플데이터 : 100명 중 혈액형 인원 비교
# --------------------------------------------

blood_type=['A','B','O','AB']
count=[45,35,15,5]

# --------------------------------------------
# 기본 막대 그래프
# --------------------------------------------
plt.bar(blood_type,count)
pass
```

```python
# --------------------------------------------
# 샘플데이터 : 100명 중 혈액형 인원 비교
# --------------------------------------------

blood_type=['A','B','O','AB']
count=[45,35,15,5]

# --------------------------------------------
# 막대그래프 옵션 저정
# --------------------------------------------
plt.bar(blood_type,count, 
        width=0.5, 
        color='pink')
pass
```

```python
# --------------------------------------------
# 샘플데이터 : 100명 중 혈액형 인원 비교
# --------------------------------------------

blood_type=['A','B','O','AB']
count=[45,35,15,5]

# --------------------------------------------
# 가로막대그래프, 옵션지정, 그래프 크기 설정
# --------------------------------------------
plt.figure(figsize=(4,3))
plt.barh(blood_type,count,
         color = 'gray',
         height = 0.5)
pass
```

## 파이그래프
* 비율의 시각화
* **`plt.pie(데이터)`**
>* 주요 옵션
>   * `labels` : 각 조각에 표시할 라벨 리스트
>   * `autopct` : 퍼센트 표시 형식 (예:'%.1f%%')
>   * `colors` : 조각 색상 리스트
>   * `explode` : 각 조각을 얼마나 띄울지 결정
>   * `wedgeprops` : 각 부채꼴 조각의 스타일 지정

```python
# --------------------------------------------
# 샘플데이터 : 100명 중 혈액형 인원 비교
# --------------------------------------------
blood_type=['A','B','O','AB']
count=[45,35,15,5]

# --------------------------------------------
# 기본 파이그래프
# --------------------------------------------
plt.pie(count, labels=blood_type)
plt.show()
```

```python
# --------------------------------------------
# 샘플데이터 : 100명 중 혈액형 인원 비교
# --------------------------------------------
blood_type=['A','B','O','AB']
count=[45,35,15,5]

# --------------------------------------------
# 파이그래프 옵션지정
# labels
# autopct
# colors = ['#E74C3C','#3498DB','#9B59B6','#2ECC71']
# explode = [0,0,0,0.3]
# wedgeprops = {'width':0.7, 'edgecolor':'white'}
# --------------------------------------------
plt.pie(count, 
        labels=blood_type,
        autopct="%.2f%%",
        explode=[0,0,0,0.1],
        wedgeprops = {'width':0.7, 'edgecolor':'w'},
        shadow=True,
        colors = ['#E74C3C','#3498DB','#9B59B6','#2ECC71'])
pass
```

## 중첩플롯
* 하나의 axes 위에 여러 개의 그래프를 겹쳐서 시각화하는 방식
* 여러 변수 간의 비교나 추이를 한 눈에 파악할 수 있다.
* 하나의 코드셀에 그래프를 여러개 그리면 된다.
* 각 그래프를 구별하기 위한 범례를 표시한다.
    * 범례를 표시하는 방법
        * 그래프 label 옵션 지정
        * `plt.legend()` 설정

```python
# --------------------------------------------
# 샘플 데이터 
# 두 사람의 월별 몸무게 변화
# --------------------------------------------

df1 = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[80,78,75,73,70]})
df1 = df1.astype({"월":"str"})
df2 = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[60,62,59,55,54]})
df2 = df2.astype({"월":"str"})
```

```python
df1
```

```python
df2
```

```python
# --------------------------------------------
# 기본 중첩플롯
# --------------------------------------------
plt.plot(df1['월'], df1['몸무게'], label='철수')
plt.plot(df2['월'], df2['몸무게'], label='영희')
plt.legend()
pass
```

```python
# --------------------------------------------
# 중첩플롯 - 범례 표시
# --------------------------------------------
plt.plot(df1['월'], df1['몸무게'], label='철수', marker="o")
plt.plot(df2['월'], df2['몸무게'], label='영희', marker="^")
plt.legend()
plt.grid()
plt.xlabel("월")
plt.ylabel("몸무게(kg)")
plt.title("월별 몸무게 변화")
plt.show()
```
