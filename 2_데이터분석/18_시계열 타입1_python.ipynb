## 날짜 시간 포메팅
* `datetime.datetime.strftime(포멧)`
    - string format time
    - 지정한 포멧의 **문자열로** 반환
    - https://strftime.org/

```python
print("두자리 연도\t('%y') :" )      
print("네자리 연도\t('%Y') :" )
print("월이름 FULL\t('%B') :" )
print("월이름 약자\t('%b') :" )
print("월 zero padding\t('%m') :" )
print("일 zero padding\t('%d') :" )
print("시 00~23 \t('%H') :" )
print("시 00~12 \t('%I') :" )
print("AM, PM \t\t('%p') :" )
print("요일명 FULL\t('%A') :" )
print("요일명 약자\t('%a') :" )
print("요일번호(0:Sun~)\t('%w') :" )
```
