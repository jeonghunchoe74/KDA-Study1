## 앤스콤 데이터셋

```python
# --------------------------
# 앤스콤 데이터셋
#   - 1973년 통계학자 프랜시스 앤스콤이 만든 데이터셋
#   - 시각화의 중요성을 보여주기 위해 설계됨
#   - 각 데이터셋은 거의 동일한 기본 통계값(갯수, 평균, 표준편차)을 가지고 있지만
#     시각화했을 때는 와전히 다른 패턴을 보여줌
# --------------------------

import seaborn as sns
anscombe = sns.load_dataset('anscombe')

df1 = anscombe[anscombe['dataset']=='I']
df2 = anscombe[anscombe['dataset']=='II']
df3 = anscombe[anscombe['dataset']=='III']
df4 = anscombe[anscombe['dataset']=='IV']
df1.head()
```

```python
# --------------------------
# 앤스콤 데이터셋 통계수치 확인
# --------------------------

print(f'df1 >>>>> mean{df1['y'].mean().round(2)}:, std:{df1['y'].std().round(2)}')
print(f'df2 >>>>> mean{df2['y'].mean().round(2)}:, std:{df2['y'].std().round(2)}')
print(f'df3 >>>>> mean{df3['y'].mean().round(2)}:, std:{df3['y'].std().round(2)}')
print(f'df4 >>>>> mean{df4['y'].mean().round(2)}:, std:{df4['y'].std().round(2)}')
```

## 서브플롯 - 방식1
* `plt.subplot(nrows, ncols, index)`
    * nrows : 서브플롯 행의 수
    * ncols : 서브플롯 열의 수
    * index : 순서

```python
# --------------------------
# 서브플롯의 기본 틀
# --------------------------

plt.subplot(2,2,1)
# 그래프1

plt.subplot(2,2,2)
# 그래프2 

plt.subplot(2,2,3)
# 그래프3

plt.subplot(2,2,4)
# 그래프4

pass
```

```python
# --------------------------
# 서브플롯 채우기
# --------------------------

plt.subplot(2,2,1)
plt.plot(df1['x'], df1['y'], 'o')

plt.subplot(2,2,2)
plt.plot(df2['x'], df2['y'], 'o')

plt.subplot(2,2,3)
plt.plot(df3['x'], df3['y'], 'o')

plt.subplot(2,2,4)
plt.plot(df4['x'], df4['y'], 'o')

pass
```

```python
# --------------------------
# 각 서브플롯에 제목 지정
# --------------------------

plt.subplot(2,2,1)
plt.plot(df1['x'], df1['y'], 'o')
plt.title('I')

plt.subplot(2,2,2)
plt.plot(df2['x'], df2['y'], 'o')
plt.title('II')

plt.subplot(2,2,3)
plt.plot(df3['x'], df3['y'], 'o')
plt.title('III')

plt.subplot(2,2,4)
plt.plot(df4['x'], df4['y'], 'o')
plt.title('IV')

pass
```

```python
# --------------------------
# 그래프의 간격 조정
# tight_layout()
# --------------------------

plt.subplot(2,2,1)
plt.plot(df1['x'], df1['y'], 'o')
plt.title('I')

plt.subplot(2,2,2)
plt.plot(df2['x'], df2['y'], 'o')
plt.title('II')

plt.subplot(2,2,3)
plt.plot(df3['x'], df3['y'], 'o')
plt.title('III')

plt.subplot(2,2,4)
plt.plot(df4['x'], df4['y'], 'o')
plt.title('IV')

plt.tight_layout()

pass
```

```python
# --------------------------
# 전체 그래프의 크기 조정
# plt.figure(figsize=(x,y))
# --------------------------
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.plot(df1['x'], df1['y'], 'o')
plt.title('I')

plt.subplot(2,2,2)
plt.plot(df2['x'], df2['y'], 'o')
plt.title('II')

plt.subplot(2,2,3)
plt.plot(df3['x'], df3['y'], 'o')
plt.title('III')

plt.subplot(2,2,4)
plt.plot(df4['x'], df4['y'], 'o')
plt.title('IV')

plt.tight_layout()

pass
```

```python
# --------------------------
# 전체 그래프의 제목 설정
# plt.suptitle()
# --------------------------
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.plot(df1['x'], df1['y'], 'o')
plt.title('I')

plt.subplot(2,2,2)
plt.plot(df2['x'], df2['y'], 'o')
plt.title('II')

plt.subplot(2,2,3)
plt.plot(df3['x'], df3['y'], 'o')
plt.title('III')

plt.subplot(2,2,4)
plt.plot(df4['x'], df4['y'], 'o')
plt.title('IV')

plt.suptitle('앤스콤 4분할 그래프')
plt.tight_layout()

pass
```

## 서브플롯  - 방식2
* `fig, axes = plt.subplot(nrows, ncols)`
>* 주요 옵션
>    * `figsize` : figure의 크기
>    * `shareX` : x축 눈금 공유
>    * `shareY` : y축 눈금 공유
<br>
>* 그래프 작성
>   * `axes[행][열].plot()`
>   * `axes[행][열].set_title()`
>   * `axes[행][열].set_xlabel()`
>   * `axes[행][열].set_xlim()`
<br>
>* 그래프 설정
>   *  `plt.tight_layout()` : 그래프가 겹쳐지지 않도록 설정



```python
# --------------------------
# 기본 서브플롯
# --------------------------

fig, axes = plt.subplots(2,2)

axes[0][0].plot(df1['x'], df1['y'], 'o')
axes[0][1].plot(df2['x'], df2['y'], 'o')
axes[1][0].plot(df3['x'], df3['y'], 'o')
axes[1][1].plot(df4['x'], df4['y'], 'o')

pass
```

```python
# --------------------------
# 각 서브플롯의 제목 지정
# set_title()
# --------------------------

fig, axes = plt.subplots(2,2)

axes[0][0].plot(df1['x'], df1['y'], 'o')
axes[0][1].plot(df2['x'], df2['y'], 'o')
axes[1][0].plot(df3['x'], df3['y'], 'o')
axes[1][1].plot(df4['x'], df4['y'], 'o')

axes[0][0].set_title('I')
axes[0][1].set_title('II')
axes[1][0].set_title('III')
axes[1][1].set_title('IV')
pass
```

```python
# --------------------------
# 그래프의 간격 조정
# tight_layout()
# --------------------------

fig, axes = plt.subplots(2,2)

axes[0][0].plot(df1['x'], df1['y'], 'o')
axes[0][1].plot(df2['x'], df2['y'], 'o')
axes[1][0].plot(df3['x'], df3['y'], 'o')
axes[1][1].plot(df4['x'], df4['y'], 'o')

axes[0][0].set_title('I')
axes[0][1].set_title('II')
axes[1][0].set_title('III')
axes[1][1].set_title('IV')

fig.tight_layout()
pass
```

```python
# --------------------------
# flatten()
# axes를 1차원 배열로 변환하면 더 간편하게 서브플롯을 생성 가능
# --------------------------

fig, axes = plt.subplots(2,2)
axes_flat = axes.flatten()

print('fig:',fig)
print('axes:',axes)
print('axes_flat', axes_flat)

axes_flat[0].plot(df1['x'], df1['y'], 'o')
axes_flat[1].plot(df2['x'], df2['y'], 'o')
axes_flat[2].plot(df3['x'], df3['y'], 'o')
axes_flat[3].plot(df4['x'], df4['y'], 'o')

axes_flat[0].set_title('I')
axes_flat[1].set_title('II')
axes_flat[2].set_title('III')
axes_flat[3].set_title('IV')

fig.tight_layout()
pass
```

```python
# --------------------------
# flatten()
# axes를 1차원 배열로 변환하면 더 간편하게 서브플롯을 생성 가능
#  - 반복문 용
# --------------------------

fig, axes = plt.subplots(2,2)
axes_flat = axes.flatten()

dfs = [df1, df2, df3, df4]
titles = ['I', 'II', 'III', 'IV']


for i, df in enumerate(dfs):
    axes_flat[i].plot(df['x'], df['y'], 'o')
    axes_flat[i].set_title(titles[i])

fig.tight_layout()
pass
```

```python
# --------------------------
# flatten()
# axes를 1차원 배열로 변환하면 더 간편하게 서브플롯을 생성 가능
#   - 반복문 사용 
# --------------------------

fig, axes = plt.subplots(2,2)
axes_flat = axes.flatten() 

dfs = [df1, df2, df3, df4]
for i, df in enumerate(dfs):
    axes_flat[i].plot(df['x'], df['y'], 'o')

pass
```

```python
# --------------------------
# 축 공유
# sharex
# sharey
# --------------------------

fig, axes = plt.subplots(2,2,sharex=True, sharey=True)
axes_flat = axes.flatten() 

dfs = [df1, df2, df3, df4]
for i, df in enumerate(dfs):
    axes_flat[i].plot(df['x'], df['y'], 'o')

pass
```
