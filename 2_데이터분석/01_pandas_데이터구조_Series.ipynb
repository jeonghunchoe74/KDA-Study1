### Series 객체 생성     
* **`pd.Series()`**
>* 주요 파라미터
>  * data
>    * 시리즈에 저장할 데이터. 
>    * 리스트, 튜플, 넘파이배열, 딕셔너리, 스칼라(단일 값) 모두 가능
>  * index
>    * 데이터에 연결할 인덱스(라벨)
>    * 생략하면 기본적으로 RangeIndex(0, 1, 2, ...)로 설정됨
>  * dtype
>    * 데이터 타입 지정 (예:'int', 'float', 'str', 'category' 등)
>    * 생략하면 data에서 자동 추론
>  * name 
>    * 시리즈의 이름
>    * 데이터프레임으로 변환할 때 컬럼명이 됨


```python
# ---------------------------
# 빈 시리즈
# ---------------------------
pd.Series()
```

```python
# ---------------
# 스칼라로 생성
# ---------------

data = 1
pd.Series(data)
```

```python
# ---------------
# 리스트로 생성
# ---------------

data = [100,50,30,10]
s = pd.Series(data)
print(s)
```

```python
# ---------------
# 넘파이 배열로 생성
# 인덱스라벨  : ['A', 'B', 'O', 'AB']
# 시리즈 이름 : 혈액형 
# 시리즈 데이터 타입  : float
# ---------------

data = pd.array([100,50,30,10])
idx = ['A', 'B', 'O', 'AB']
s = pd.Series(data, index = idx, name="혈액형", dtype="float")
s
```

```python
# ---------------
# 딕셔너리로 생성
# 시리즈 이름 : 혈액형 
# ---------------

data = {'A':100, 
        'B':50, 
        'O':30, 
        'AB':10}
s = pd.Series(data, name="혈액형")
s['A']   #s[0]
s.iloc[0]
s
```

### Series 데이터 접근

* 라벨 인덱스로 접근
  * **`시리즈객체.loc[인덱스라벨]`**
  * **`시리즈객체.loc[인덱스라벨 리스트]`** 
  * **`시리즈객체.loc[인덱스라벨 슬라이스]`** --> 끝인덱스의 데이터 포함
* 정수형 인덱스로 접근
  * **`시리즈객체.iloc[RangeIndex]`**
  * **`시리즈객체.iloc[RangeIndex리스트]`** 
  * **`시리즈객체.iloc[RangeIndex슬라이스]`**

```python
# ---------------------
# A형은 몇명인가요?
# 인덱스라벨 사용
#
# 결과
# np.int64(100)
# ---------------------
s.loc['A']
```

```python
# ---------------------
# A형은 몇명인가요?
# RangeIndex 사용
#
# 결과
# np.int64(100)
# ---------------------

s.iloc[0]
```

```python
# ---------------------
# A형, O형은 각각 몇명인가요?
# 인덱스라벨리스트 사용
# ---------------------

s.loc[["A","O"]]
```

```python
# ---------------------
# A형, O형은 각각 몇명인가요?
# RangeIndex리스트 사용
# ---------------------
s.iloc[[0,2]]
```

```python
# ---------------------
# A형부터 O형까지 각각 몇명인가요?
# 인덱스라벨슬라이스 사용
# ---------------------

s.loc['A':'O']
```

```python
# ---------------------
# A형부터 O형까지 각각 몇명인가요?
# RangeIndex슬라이스 사용
# ---------------------

s.iloc[:3]
```

### Series 데이터 추가, 수정, 삭제
* 데이터 추가/수정
  * `시리즈객체[인덱스] = 데이터`
* 데이터 삭제
  * `del 시리즈객체[인덱스]`

```python
# ---------------------
# 데이터 추가
# 'A-'  3
# ---------------------

s["A-"] = 3
s
```

```python
# ---------------------
# 데이터 수정
# 'A-'  5
# ---------------------
s['A-'] = 5
s
```

```python
# ---------------------
# 데이터 삭제
# 'A-'  
# ---------------------
del s["A-"]
s
```
