## 분석

#### 사고 유형별 사고건수

```python
# -----------------------------------------
# 사고건수가 많은 순서로 정렬
# -----------------------------------------
df.sort_values("사고건수", axis=0, ascending=False, inplace=True)
df
```

```python
# -----------------------------------------
# 사고건수가 많은 순서로 정렬하여 시각화
# -----------------------------------------

df.plot( kind='bar', y="사고건수", figsize=(12, 5), color= "#ff0000", title="교통사고 유형별 분석")
pass 
```

#### 사고 유형별 사망자수

```python
# -----------------------------------------
# 사망자수가 많은 순서로 정렬
# -----------------------------------------
df
df.sort_values("사망자수", axis=0, ascending=False, inplace=True)
```

```python
# -----------------------------------------
# 사망자수가 많은 순서로 정렬하여 시각화
# -----------------------------------------
df.plot( kind='bar', y="사망자수", figsize=(12, 5), color= "#ff0000", title="교통사고 유형별 분석")
pass 
```

#### 사고 유형별 중상자수

```python
# -----------------------------------------
# 중상자수가 많은 순서로 정렬
# -----------------------------------------
df.sort_values("중상자수", axis=0, ascending=False, inplace=True)
```

```python
# -----------------------------------------
# 중상자수가 많은 순서로 정렬하여 시각화
# -----------------------------------------

df.plot( kind='bar', y="중상자수", figsize=(12, 5), color= "#b700ff", title="교통사고 유형별 분석")
pass 
```

#### 사고 유형별 사망률

```python
# -----------------------------------------
# 파생컬럼 추거
# -----------------------------------------

```

```python
# -----------------------------------------
# 사망률 높은 순으로 정렬
# -----------------------------------------

df["사망률"] = df["사망자수"] / (df.loc[:, "사망자수" : "부상신고자수"].sum(axis=1)) * 100
df.sort_values("사망률", axis=0, ascending=False, inplace=True)
```

```python
# -----------------------------------------
# 사망률 높은 순으로 정렬하여 시각화
# -----------------------------------------
df.plot(kind="bar", y = "사망률",figsize=(12,4), color="#fff000", title="교통사고 사망률")
pass
```

#### 사고 유형별 중상률

```python
# -----------------------------------------
# 파생컬럼 추가
# -----------------------------------------

```

```python
# -----------------------------------------
# 중상률 높은 순으로 정렬하여 시각화
# -----------------------------------------

df["중상률"] = df["중상자수"] / (df.loc[:, "사망자수" : "부상신고자수"].sum(axis=1)) * 100
df.sort_values("중상률", axis=0, ascending=False, inplace=True)
```

```python
# -----------------------------------------
# 중상률 높은 순으로 정렬하여 시각화
# -----------------------------------------

df.plot(kind="bar", y = "중상률",figsize=(12,4), color="#7300ff", title="교통사고 중상률")
pass
```
