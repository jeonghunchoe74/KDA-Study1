```python
import pandas as pd
```

```python
# ----------------------
# 샘플데이터
# ----------------------

d = {'이름':['스티븐 호킹','마이클잭슨','스티브잡스','로빈윌리엄스','앨빈토플러'],
 '주요경력':['이론 물리학자','가수','CEO','배우','미래학자'],
 '출생':['1942-01-08','1958-08-29','1955-02-24','1951-07-21','1928-10-04'],
 '사망':['2018-03-14','2009-06-25','2011-10-05','2014-08-11','2016-06-27'],
 }
 
df = pd.DataFrame(d)
df
```

```python
# ----------------------
# dtypes 확인
# ----------------------
df.dtypes
```

### 시계열 자료형으로 변환
* `pd.to_datetime(컬럼)`

```python
# ----------------------
# 출생, 사망 컬럼
#   ===> datetime 형으로 변환
# ----------------------
df['출생'] = pd.to_datetime(df['출생'])
df['사망'] = pd.to_datetime(df['사망'])
```

```python
# ----------------------
# dtypes 확인
# ----------------------

df.dtypes
```

## 시계열 정보 추출
* `컬럼.dt.year`
* `컬럼.dt.month`
* `컬럼.dt.day`
* `컬럼.dt.quarter`
* `컬럼.dt.dayofweek`
* `컬럼.dt.day_name()`

```python

df['출생연도'] = df['출생'].dt.year
df['사망연도'] = df['사망'].dt.year
df
```

```python
df['출생'].dt.day_name()
```
