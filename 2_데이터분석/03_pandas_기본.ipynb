### 조건으로 데이터 선택
* `데이터프레임.loc[조건]`
* 비교연산
  * 비교연산자는 기본 파이썬의 비교연산자와 동일하다 
    * `>`, `<`, `>=`, `<=`, `==`, `!=`
* 논리연산
  * 여러 조건을 결합하는 경우 각 조건은 소괄호로 묶는다.
  * 논리연산자는 기호를 사용한다.
    * `&` (and)
    * `|` (or)
    * `~` (not)
* 멤버연산
  * `.isin(value리스트)`
* null여부
   * `.isnull()`
   * `.notnull()`

```python
# ------------------------------------
# 국어점수가 100점인 학생의 모든 데이터 
# ------------------------------------
cond = (df["국어"] == 100)



```

```python
# ------------------------------------
# 국어점수가 100점인 학생의 이름, 성별, 국어 
# ------------------------------------

df.loc[cond, ["이름", "성별", "국어"]]
```

```python
# ------------------------------------
# 국어,영어,수학 모두 100점인 학생의 데이터
# -------------------------------------

cond1 = (df["국어"] == 100) 
cond2 = (df["영어"] == 100)
cond3 = (df["수학"] == 100)
df.loc[cond1 & cond2 & cond3] 

```

```python
# ------------------------------------
# 국어,영어,수학 중 100점이 하나라도 있는 학생의 데이터
# ------------------------------------
cond1 = (df["국어"] == 100) 
cond2 = (df["영어"] == 100)
cond3 = (df["수학"] == 100)
df.loc[cond1 | cond2 | cond3] 
```

```python
# ------------------------------------
# 국어점수가 50점, 100점인 학생의 데이터
# ------------------------------------
cond = df["국어"] == 50 
cond1 = df["국어"] == 100
df.loc[cond | cond1]
```

```python
# ------------------------------------
# 국어점수가 null인 학생의 데이터
# ------------------------------------

cond = df["국어"].isnull()
df.loc[cond]
```

```python
# ------------------------------------
# 국어점수가 null이 아닌 학생의 데이터
# ------------------------------------
cond = df["국어"].notnull()
df.loc[cond]
```

## 열 추가/삭제

#### 파생컬럼 추가
* **`데이터프레임[컬럼] = 추가할데이터`**  
* **`데이터프레임[컬럼] = 수정할데이터`**  
* 컬럼이 존재하면 수정, 존재하지 않으면 추가된다.
* 컬럼 간 연산 시 NaN이 존재하는 경우 NaN이 전파된다.
* 판다스의 집계함수(sum, mean 등)를 사용하는 경우 NaN을 무시하고 존재하는 데이터를 이용하여 계산됨.


```python
# --------------------------------------
# 평균 칼럼 추가 
# --------------------------------------

df["평균"] = df.loc[:, "국어" : "수학"].mean(axis=1).round(2)
df
```

#### 열 삭제
* `데이터프레임.drop(컬럼라벨, axis=1)`

```python
# --------------------------------------
# 평균 칼럼 삭제 
# --------------------------------------
df = df.drop('평균', axis=1)
```

#### 결측치 확인

  * **`데이터프레임객체.isnull().sum()`**

#### 결측치 처리
* 결측치가 포함된 행 삭제
  * **`데이터프레임객체.dropna()`**  : 모두 삭제
  * **`데이터프레임객체.dropna (subset=컬럼리스트)`** : 해당 컬럼만 삭제
* 결측치를 다른 값으로 채우기
  * **`데이터프레임객체.fillna(값)`** 채우기
  * **`데이터프레임객체[컬럼].fillna(값)`**
  * **`데이터프레임.ffill()`** : 이전값으로 채우기
  * **`데이터프레임.bfill()`** : 다음값으로 채우기 
  * **`데이터프레임.fillna({'컬럼명1':값1, '컬럼명2':값2,...})`** : 컬럼별로 값을 지정하여 채우기

```python
# --------------------------------------
# 결측치를 0으로 채우기
# --------------------------------------
df.fillna(0, inplace=True)
df
```
