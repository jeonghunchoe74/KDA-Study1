#### `컬럼.rank()`
>* 주요 파라미터
>* method
>    * `'average'`: 동점인 경우 평균순위 부여
>    * `'min'` : 동점인 경우 최소 순위 부여
>    * `'max'` : 동점인 경우 최대 순위 부여
>    * `'first'` : 데이터 등장 순서대로 부여
>    * `'dense'` : min과 유사하지만 순위 간격 없이 연속적으로 부여
>* ascending
>    * `True` : 오름차순 순위
>    * `False` : 내림차순 순위
