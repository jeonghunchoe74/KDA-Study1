### 데이터프레임 생성

```python
# ------------------------------------------------
#  파일경로 패턴을 만족하는 여러 파일 리스트를 불러온다.
# ------------------------------------------------

import glob
file_list = glob.glob('data/LOCAL_PEOPLE_DONG/LOCAL*.csv')
file_list
```

```python
# ------------------------------------------------
#  파일 리스트의 데이터들을 데이터프레임으로 각각 변환하여
#  데이터프레임 리스트에 추가한다.
# ------------------------------------------------

df_list = []

for file in file_list:
    df_month = pd.read_csv(file,
                     index_col=False,   # 첫번째 컬럼이 인덱스로 자동 지정되는 문제 해결
                     parse_dates = ['기준일ID'],    # datetime 자료형으로 지정할 컬럼
                     dtype={'행정동코드':'str'})    # object 자료형으로 지정할 컬럼
    df_list.append(df_month)
```

```python
print('len:', len(df_list))
```

```python
for df in df_list:
    display(df.head(2))
```

```python
# -------------------------------------------------------------
# 데이터프레임 리스트의 데이터프레임을 하나의 데이터프레임으로 합친다.
# -------------------------------------------------------------

df_all = pd.concat(df_list, ignore_index=True)
```

### 월, 일 컬럼 추가

```python
# ---------------------------------------
# 월, 일 컬럼 추가
# ---------------------------------------

df_all['월'] = df_all['기준일ID'].dt.month
df_all['일'] = df_all['기준일ID'].dt.day
```

### 요일 컬럼 추가
* `dt.dayofweek` : 요일번호(0:Mon~)
* `dt.day_name()` : 요일명 full name (Monday, Tuesday, ...)
* `dt.day_name().str[:3]` : 요일명 약자 (Mon, Tue, ...)
* `dt.strftime('%A')` : 요일명 full name (Monday, Tuesday, ...)
* `dt.strftime('%a')`  : 요일명 약자 (Mon, Tue, ...)

```python
# ---------------------------------------
# 요일 컬럼 추가
#   day_name()
# ---------------------------------------

df_all['요일'] = df_all['기준일ID'].dt.day_name().str[:3] 
```

### 시도명, 시군구명, 행정동명 추가

#### [참고] 법정동과 행정동          
    * 법정동과 행정동은 우리나라의 행정구역을 구분하는 두 가지 서로 다른 체계입니다.

---

* 법정동     
    * 법률로 지정된 공식 행정구역
    * 1914년 행정구역 통폐합 때 만들어져 현재까지 유지
    * 국토교통부에서 관리
    * 등기부등본, 신분증 등 모든 법적 문서의 주소 표기에 사용
    * 부동산 규제 등 법적 업무의 기준
    * 명칭 변동이 거의 없음

---

* 행정동
    * 행정 편의를 위해 설정된 관리 구역
    * 행정안전부에서 관리
    * 인구수와 행정 수요에 따라 수시로 조정 가능
    * **행정동 코드는 자주 변경되므로, 해당 데이터의 행정동 코드 기준 시점을 확인하여야 함.**
    * **생활인구 데이터는 데이터 일관성 유지를 위해 2016년 집계구, 행정동 코드를 사용하고 있음**
   
---
* 법정동과 행정동의 관계
    * 두 구역은 N:M 관계
    * 하나의 법정동이 여러 행정동으로 나뉠 수 있음
    * 여러 법정동이 하나의 행정동으로 통합될 수 있음
    * 일부 지역은 법정동과 행정동이 일치하기도 함
    * 예를 들어, 서울 중구의 경우 법정동은 87개지만 행정동은 17개만 존재
    * 반대로 송파구 잠실동(법정동)은 잠실본동, 잠실2동, 잠실3동, 잠실7동이라는 4개의 행정동으로 나뉨

#### 행정동코드 매핑정보 
* 행정구역코드정보 파일 다운로드
    *  https://data.seoul.go.kr/dataVisual/seoul/seoulLivingPopulation.do

```python
# -----------------------------------------------
# 행정동코드와 행정동명 매핑을 위한 데이터프레임 생성
# -----------------------------------------------

file_path = 'data/LOCAL_PEOPLE_DONG/행정동코드_매핑정보_20241218.xlsx'

df_region = pd.read_excel(file_path, 
                          sheet_name='행정동코드', # 엑셀파일의 시트명
                          header = 0,             # 헤더(컬럼인덱스)
                          skiprows = [1],         # 불러오지 않을 행
                          usecols = [1,2,3,4],    # 사용할 컬럼
                          dtype = {'행자부행정동코드':'str'}) # 데이터타입
df_region.head()
```

```python
# -----------------------------------------------
# 일관성을 위해 컬럼명 변경
# -----------------------------------------------

df_region.rename({'행자부행정동코드':'행정동코드'}, 
                 axis=1, 
                 inplace=True)
```

```python
df_region.head()
```

```python
# -----------------------------------------------
# 시도명, 시군구명, 행정동명 추가
# -----------------------------------------------

df_all = pd.merge(df_all, df_region)
df_all.head()
```
